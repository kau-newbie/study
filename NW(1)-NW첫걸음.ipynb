## **| 우리 집 NW 들여다보기**

!참고

CPE(Custom Premises Equipment): 고객 집 안에 설치되는 모뎀 등의 장치. 회사의 NW와 집안의 NW를 연결해준다.

#### **NW구성요소**

- 컴퓨터(PC나 서버)
- NW장비(라우터나 스위치 등)
- 전송 매체(케이블, 전파 등)

> 보통 application이 데이터를 송수신하는 주체. 컴퓨터에 들어있다.

NW장비는 app의 데이터를 올바르게 전송하기 위한 기기. 

> - **라우터** : NW를 구분 짓고 서로 연결해주는 장비.
>
>   기술적인 의미로 NW란, '라우터에 의해 구분 지어지는 범위'를 가리킨다.
>
> - **스위치** : 같은 NW 내부의 데이터 전송을 수행하는 장비.
>
>   스위치가 NW의 입구에 해당하는 것.

#### **홈 NW 구성예시**

가정집 내부의 NW(LAN). 
1. PC나 프린터, TV
2. 브로드밴드 라우터(인터넷 연결 기능을 가진 개인용 라우터)나 스위치
3. LAN 케이블

브로드 밴드 라우터의 포트에는 LAN케이블이 꽂혀있지만, NW 바깥으로 향하는 방향으론 WAN 포트가 있다. (내부는 반대로 LAN 포트)

--> 브로드밴드 라우터의 LAN포트 쪽이 하나의 NW에 해당한다.

홈NW는 말 그대로 집 안의 기기들끼리만 통신이 가능하다.

-> 홈 NW를 인터넷과 연결해야한다.

-> **ISP**등장 (Internet Service Provider)

인터넷 접속을 제공하는 기업을 말한다. 전 세계 ISP들이 서로 연결된 NW가 바로 인터넷.

ISP에 따라 제공되는 SV가 달라진다.

1. ADSL
2. FTTH
3. CATV
4. WiMAX

홈NW를 인터넷에 연결하는 데에는 브로드밴드 라우터를 사용하지만,

위 방식에 따라서 브로드밴드 라우터의 **바깥쪽에 연결할 장비**(고객 집 안의 장치)와 **전송 매체**(케이블)이 다름.

|SV|고객 집 안 장치|케이블| |
|---|---|---|---|
|   |   |**브로드밴드 라우터와 집 안 장치사이**| **집안 장치의 바깥쪽**|
|ADSL| ADSL modem|LAN 케이블(공통)|전화선|
|FTTH|ONU(Optical NW Unit)| | 광섬유|
|CATV|CATV modem|   |동축케이블|

- 고객 집 안 장치는 물리적 신호를 변환하는 역할을 함.

<-- 전송 매체를 오가는 물리적 sig와 ISP 사이의 전송 매체를 오가는 신호가 서로 다르기 때문. (서로 다른 sig.상호 변환)

- 고객 집 안 장치는 브로드밴드 라우터의 WAN 포트에 LAN 케이블로 연결함.

- 브로드밴드 라우터와 ISP의 라우터 사이가 하나의 네트워크에 해당함.

ISP --> 라우터 -->-(WAN 케이블)- 고객 댁네 장치(modem) -(LAN 케이블)- --> 브로드밴드 라우터

**즉, 브로드밴드 라우터 - ISP라우터의 네트워크 한 개와,**

**다시 브로드밴드 라우터 - 집안 디바이스들의 네트워크 한 개**, 

**총 두 개가 브로드밴드 라우터로 이어짐.**

**LAN**


LAN은 지리적 좁음 보다는,

1. 사용자가 직접 관리, 구축한다.
2. 통신요금이 필요 없다.

LAN의 규격

>LAN 구축 위해선 라우터나 스위치 등 장비와 LAN 케이블 필요.

1. 이더넷
2. 토큰링
3. FDDI

Ethernet이 가장 보급된 LAN규격. 나머지 두 개는 효율과 신뢰성 위해 복잡한 장비, 비용듬.

|규격 명|케이블|전송속도|
|---|---|---|
|10BASE5|동축 케이블|10Mbps|
|100BASE-T@|카테고리 x 이상의 UTP 케이블| 100Mbps|
|1000BASE-@X|광섬유|1Gbps(1000Mbps)|

이더넷은 다양한 전송 속도가 있지만 기본적 구조는 거의 같음.


데이터 전송 범위는 같은 NW 내부 한정.


**데이터를 전송하기 위한 주소로 MAC주소를 사용함**

#### **인터넷이란**

> LAN의 바깥 세계

인터넷은, **전 세계의 ISP가 서로 연결된 NW**

ISP는 계층 구조로 이루어져있고, 일정 규모이상에 도달하면 **Tier1 ISP**이라고 부름. 

(ISP를 전문적으로 AS- Autonomous Sys.라고도 부름. 정확히 ISP는 AS의 일종)

Tier1 ISP -> ISP -> 가정집 NW

아무튼, *인터넷에 연결한다는 것은*, *ISP와 연결한다*는 것.

ISP와 연결하기 위한 통신 회선을 '엑세스 회선'이라고 부름. 

**<엑세스 회선>**


|고정회선|모바일 회선|
|---|---|
|전용 회선|PHS회선|
|광섬유|핸드폰 회선|
|케이블 TV회선|WiMAX회선|

대기업은 전용 회선을 통해 신뢰도를 확보.


<-->

보통 저렴하고 QoS가 상대적으로 낮은 브로드밴드 인터넷SV를 이용함.

LTE, 5G, WiMAX 회선 등의 모바일 회선은 '모바일 브로드밴드 SV'

**인터넷과 WAN의 차이**

규모가 큰 회사예시:

각 거점들(LAN)끼리의 통신도 필요하다.

--> WAN 등장! LAN과 달리, 전기통신사업자가 관리해줌. 대신 돈 받으.ㅁ

`LAN`과 `WAN`, *둘 사이의 차이는*

**NW 관리자와 통신상대**

그럼,

`WAN`과 `인터넷`의 차이?

인터넷 NW는 전기통신사업자가 모두 관리하지 X. 

--> 관리는 각 접속 사업자인 ISP들이 각각 분담, 관리 범위도 각자의 것.

또, WAN도 사내 NW의 일부에 해당함.

**인터넷은 외부 NW이다.**

--> 기업간 통신에서도 요즘 인터넷을 사용.

--> 보안 이슈 발생

--> **인터넷 VPN을 사용함.**

**protocol과 NW architecture**

LAN이나 인터넷은 데이터 전송을 위한 infra.

NW를 통해 데이터를 주고받는 주체는 PC등에서 동작하는 app.

app끼리 데이터를 주고받기 위한 규약이 필요함

**프로토콜**

> - 데이터의 포맷
> - 데이터를 송수신 하는 순서
> - 데이터의 송신처
> - 에러 처리

**OSI 참조 모델** - 7계층이라 복잡해서 잘 쓰진 않음.

다만, 아키텍쳐 모델로 쓰기보단, NW 기능이나 장비의 특징, NW상태를 언급할 때 사용되고 있음.

예시: L3스위치, L4-7스위치 등등

대신, 오늘날

**TCP/IP모델**

|TCP/IP|protocol|
|---|---|
|응용 계층|HTTP, FTP, DNS, DHCP, SNMP, SMTP 등|
|전송 계층|TCP,UDP 등|
|인터넷 계층|IP, ARP, ICMP 등|
|NW/IF 계층|이더넷, 토큰링, ATM, FDDI 등|

**역할:** 4개 계층의 다양한 프로토콜을 조합해 통신을 구현. 

--> 4개 계층은 app끼리 데이터를 주고받도록 저마다 규약을 규정.

응용 계층은 데이터 포맷이나순서, 그 밑은 데이터를 목적지에 올바르게 전송하기 위한 규약 등.

RFC라는 문서에 정리된, TCP와 IP를 기반으로 한 수많은 프로토콜 체계를 일컫는다.

네트워크 IF 계층의 프로토콜은 주로 LAN이나 WAN에서의 데이터를 전송하기 위한 것.

--> TCP/IP통신시, LAN이나 WAN의 프로토콜을 자유롭게 고를 수 있음.

<br>

통신 프로토콜의 계층에 따라 데이터 단위 명칭도 달라짐.

|메시지|표현계층|
|---|---|
|세그먼트|전송계층|
|데이터그램|전송/NW계층|
|패킷|NW계층|
|프레임|DL계층|
|PTR|IP주소에 해당하는 호스트명|

정리하자면,

**컴퓨터 내 목적지 app을 위한 전송** = 여러 app이 동작하고 있을 때, 목적 app에 전송되기 위한 `전송 계층`

**app을 실행하고 있는 컴퓨터끼리의 전송** = `인터넷 계층`

**같은 NW내부의 데이터 전송** = 스위치 포함, 하나의 라우터 안에서는 `NW/IF 계층`


하위 계층의 통신 경로가 기반이 된다.

> 예를 들어, PC에서 웹사이트를 볼 떄,
>
> 1. PC의 웹 브라우저와 웹서버의 웹서버 app사이의 `전송 계층`의 통신 경로가 정해짐.
> 2. 이 전송계층 통신경로는 `인터넷 계층`의 통신 경로를 기반으로 함.
> 3. 이 인터넷 계층의 통신경로는 PC와 웹서버의 NW구성에 의한, 복수의 `NW/IF 계층`통신 경로들의 조합임.

**하위 계층이 장애 --> 상위 계층 기능불가.**

#### **IP주소?**

TCP/IP로 통신하는 기기를 식별하기 위한 주소.

통신 기기들을 호스트라고 함.

호스트: PC와 서버, 라우터 등 NW장비.

--> app을 실행중인, **호스트를 특정하기 위한 정보**가 **IP주소**

IP주소는 호스트의 유선 LAN, 무선 LAN등의 NW IF(IF는 NW와 컴퓨터간의 연결지점으로, '문'이다. 랜카드, WI-FI card등이 있다.)마다 설정된다.

IP주소는...


- 32bit. --> 8bit씩 10진수로 변환해 표기.
- 유니캐스트, 브로드캐스트, 멀티 캐스트로 구분.
  > - 브로드 캐스트와 멀티 캐스트는 1:n통신.
  > - 브로드 캐스트: 같은 NW상 모든 호스트
  > - 멀티 캐스트: 어떤 특징으로 그룹화된 호스트

- **IP주소 = 네트워크 ID + host ID**

  NW ID로 NW를 식별, 뒷부분의 host ID로 네트워크 내부의 호스트(IF)를 식별한다.

단, 어디서부터 어디까지가 NW ID인지, host ID인지는 구분X

--> **class**가 등장.

#### **클래스**

IP주소는 A~E클래스로 구분.

**A클래스**

- 맨 앞 비트패턴 0
- 맨 앞 8비트 10진수 표기 1~126 (2^7)
- NW ID와 호스트 ID구분은 8비트째에서
- NW ID개수는 126개
- 호스트 ID개수는 1600만개

**B클래스**

- 맨 앞 비트패턴 10
- 맨 앞 8비트 10진수 표기는 128~191 ('10' + 2^6)
- NW ID와 host ID구분은 16비트째에서
- NW ID개수는 16,382개

**C클래스**

- 맨 앞 비트패턴 110
- 맨 앞 8비트 10진수 표기는 192~223('110' + 2^5)
- NW ID와 Host ID 구분은 24비트 째에서( 8 X 3)
- NW ID수는 209만 7,150개

이런 식이다.

클래스에 기초한 IP주소를 **클래스풀 주소** 라고 부름.

하지만 낭비가 심함.

예를 들어,
> A클래스는 NW 126개 각각에 호스트 1600만대 가능.
>
> 하지만 현실적으로 NW 하나에 1600만대? 다 남아돌지.
>
> --> **서브넷 마스크**등장( **클래스리스 주소**)

#### **서브넷 마스크**

> IP 주소와 세트로 사용.
>
> IP주소의 32bit중 **'어디까지가 NW ID, 또 어디부터가 Host ID인지'** 알림.

- 서브넷 마스크도 IP주소를, 8비트씩 10진수로 변환해 표기.
- **비트가 1 = NW ID**이고, **비트가 0 = host ID**를 나타냄
- 반드시 연속1이 끝나면 연속0이 나온다.

프리픽스 방법도 잇는데, 비트 1이 연속으로 나온 갯수를 적음.

예시: /24 <== 비트1이 연속 24(8x3)개.

**클래스리스주소**는 NWID와 host ID를 **8비트 단위로 구분하지 않음.**

#### **공인 IP주소**

인터넷 통신에서 사용하는 주소.

보통 공인 IP주소는 계약한 ISP에서 할당받게 된다.

1. 네이버에서 '내 IP검색'
2. whois에서 이 공인 IP를 복붙 후 검색.

- 공인 IP는 인터넷 상에서 중복되지 않는 addr.
- 공인 IP는 ICANN(Internet Coporation for Assigned Names and Numbers)에서 관리.

ICANN산하 `IR`에서는 다시 전 세계 지역을 나눠 담당하는 `RIR`로 나뉘고, 다시 국가별 `NIR`, 또다시 국가 안에서 `LIR(= ISP)`로 나뉜다.

#### **사설 IP**

> 공인 IP는무한하지 X.

외부와 연결되지 X는 NW에서는 공인 IP필요 X.

--> **공인 IP 주소로 이용되지 않는 범위의 Ip주소를 설정해 이 안에서 돌려씀** 이게 사설 IP.


**공인 IP와 사설 IP의 범위**

IPv4의 전체 범위인 0.0.0.0 ~ 255.255.255.255 안에서 다음과 같이 역할이 나뉩니다. 

사설 IP는 공인 IP와 별개의 체계가 아니라, 전체 IP 대역 중 일부를 사설용으로 예약해 둔 것입니다.

**공인 IP로 사용할 수 없는 주요 대역**

`사설 IP 주소 대역`

이 대역들은 내부 네트워크(가정, 회사 등)에서 자유롭게 사용하도록 약속된 주소입니다.

- A 클래스: 10.0.0.0 ~ 10.255.255.255

- B 클래스: 172.16.0.0 ~ 172.31.255.255

- C 클래스: 192.168.0.0 ~ 192.168.255.255

`루프백(Loopback) 주소`

- 127.0.0.0 ~ 127.255.255.255 대역은 컴퓨터 자기 자신을 가리키는 주소로 예약되어 있습니다. 

네트워크 장비의 테스트 등을 위해 사용되며, 특히 127.0.0.1이 대표적입니다.

`기타 특수 목적 주소`

 - 0.0.0.0: '정의되지 않은 주소'를 의미하는 특수한 주소입니다.

- 224.0.0.0 ~ 239.255.255.255: 멀티캐스트(한 번의 전송으로 특정 그룹에게만 데이터 전송)를 위해 예약된 주소입니다.

- 240.0.0.0 ~ 255.255.255.255: 미래를 위해 남겨둔 실험적인 주소 대역입니다. (단, 255.255.255.255는 브로드캐스트 주소로 사용됩니다.)

<br>

#### **NAT**

사설 IP는 닫혀있는 NW안에서 사용.

--> 따라서 인터넷상에선 목적지 주소가 사설 IP주소가 X.

( *라우터는 라우팅 테이블에서 없는 NW ID를 가진 패킷을 파기.* )

*사설 IP를 쓰는 호스트를 생각하면...*

그 NW의 라우터는 보내는 라우팅은 가능하지만,

**받을 때는 불가능**(사설 IP로 돼있기 때문.)

--> **NAT** (network address translation)이 등장!

NAT에 의한 주소 변환은 사설 IP주소와 공인 IP주소의 경계에 위치하는 라우터나 방화벽 상에서 이루어짐.

<br>

**NAT의 동작**

예시: 사설 NW의 PC와(PC1) 인터넷상의 서버(SRV1)이 통신한다고 가정.

1. PC1이 SRV1으로 보내는 IP 패킷 -

- 목적지 주소는 200.1.1.1 (공인 IP)
- 발신자 주소는 192.168.1.1(사설 IP)

--> NAT을 수행하는 라우터(R1)은 사설 IP와 공인 IP의 경계에서, 주소 192.168.1.1을 공인 IP주소 100.1.1로 변환해 전송한다.

2. 이때, R1은 변환 정보를 NAT table에 기록. 이 기록을 받을 패킷', 즉 SRV1이 PC1으로 돌려보낸 IP 패킷이 도착할 때 사용함.

이때! 라우터가 인터넷에서 없는 사설 IP로 패킷을 보내면 받아서 파기.

--> 따라서, **인터넷에서 사설쪽으로 먼저 통신을 시도할 순 XXXX**

보통 하나의 공인 IP에 여러개의 사설 IP를 짝 지어주는 확장 기능이 존재.
- NAPT
- IP Masquerade 등이 있다.

## **| DNS구조 이해하기**

DNS 서버 IP주소는 '인터넷 프로토콜 버전4(TCP/IP) 속성' 윈도우에서 설정.

> DNS는 (Domain Name Sys.) 호스트명과 IP주소 매핑을 수행하는 sys를 말한다.

이름 해석을 하는 것.

원래는 *hosts*라는 파일을 사용하다 (호스트명과 IP주소의 매핑정보를 기록한 파일), 

--> **DNS**를 이용한 이름 해석을 사용했다.

#### **리소스 레코드**

DNS로 이름 해석을 위해 DNS 서버에 host명과 Ip주소의 매핑을 `리소스 레코드`라고 부름.

- A타입~NS 타입이 있다.

#### **도메인과 존**

`도메인` : 호스트명의 집합
- 기업 등의 조직에 의해 관리된다.
- 도메인 내의 호스트에는 공통된 호스트명이 부여됨.

도메인을 또다시 서브 도메인으로 나누기 가능.

예를 들면, www.n-study.com, mail.n-study.com같이 모든 호스트명에 *n-study.com*이 붙는 식.

`존` : DNS 서버가 관리하는 범위. 
- 구체적으로 어떤 DNS서버에 등록된 리소스 레코드의 집합이 존에 해당.


```

---------dns.n-study.com이 관리하는 존-------------------------------------------------
| DNS 서버(dns.n-study.com), web srv(www.n-study.com), mail srv(mail.n-study.com) etc |
|                                                                                     |
|----------------dns.sub.n-study.com이 관리하는 존-------------------------------------|
| DNS 서버(dns.sub.n-study.com), pc(pc1.sub.n-study.com) etc                          |
|                                                                                    |
--------------------------------------------------------------------------------------


이렇게 존 안에 존, 각 도메인 안에 도메인이 존재할 수 있다.

**도메인의 계층**

인터넷에 연결된 다양한 조직들의 NW.

각 조직에는 많은 수의 host.

다양한 host들의 효율적 통신을 위해 계층을 나눔.

예시: 

계층 최상위에 **루트를 정점으로 한** `트리구조.`

1. 루트 하위계층에 `'com'`, `'kr'`, `'org'`등의 도메인.

2. `'kr'` 하위계층에 `'co'`, `'ac'`, `'go'` 등의 도메인.

--> 이런 각 도메인에는 **DNS 서버**가 설치된다.

이때, 각 DNS 서버는 **루트 DNS 서버의 IP주소**와 **하위 계층 도메인 NDS서버의 IP주소**를 알고 있다.

**DNS 이름 해석**

> DNS 서버에 네임 쿼리(DNS 서버에 전송하는 호스트명 질의문)를 송신해, 목적지 호스트명에 매핑된 IP주소를 반환함.

만일, 해당하는 **매핑정보가 없으면**, **루트로부터 도메인 계층구조를 따라가며 질의**를 계속함.

--> `com`을 찾고, 그다음 `naver`을 찾는 식.

- DNS 서버에 질의를 보내는 클라이언트를 resolver라고 부른다.

dns 서버나 resolver는 cache형태로 maping정보를 저장. `ipconfig /displaydns`를 실행시켜보자. 

대부분의 경우, **홈NW**에 속한 PC에는 **브로드밴드 라우터의 IP주소**가 **DNS 서버 주소**로 지정됨.

그리고 **브로드밴드 라우터의 DNS 서버**로는 계약한 **ISP의 DNS 서버**가 지정됨.

<br>

### **인터넷상의 서버 IP주소를 확인.**

인터넷상에 공개된 DNS 서버에 질의하면, 인터넷에 연결된 다양한 웹 서버들의 IP주소를 알아낼 수 있다. 

`nslookup`명령어사용.

보통 DNS질의는 자동.

단, 문제 생기면 수동으로. 그게 'nslookup'명령어.

#### **A레코드 질의**

**리소스 레코드**에는 A, AAAA, CNAME, MX, NS, PTR 등의 종류가 있다.

A레코드는 호스트명에 매핑된 IP주소를 의미한다. (AAAA는 ipv6)

그냥 찾고싶은 호스트명만 입력하면됨.

```bash

>
C:\Windows\System32>nslookup
기본 서버:  dns.google
Address:  8.8.8.8

> www.daum.net
서버:    dns.google
Address:  8.8.8.8

권한 없는 응답:
이름:    daum-4vdtymgd.kgslb.com
Address:  211.249.220.24
Aliases:  www.daum.net

>

- 권한이 없는 응답이란, *질의를 보낸 서버*가 **아니라**,

**다른 DNS 서버**에 등록돼있던 데이터라는 것.

- 이름은 질의 대상인 웹 서버의 호스트명.

- aliases는 웹 서버의 별칭.

#### **PTR 레코드 질의**

>PTR 레코드:
>
>IP주소에 매핑된 호스트명의 정보.

**서버**가 클라이언트로부터 어떤 요청을 받았을 때, 

클라이언트의 IP주소는 알아도, *호스트명을 모르는 경우*가 있음.

--> 이럴 때, PTR 레코드를 질의하면, IP주소에 매핑된 호스트명의 정보를 알아낼 수 있음.

IP주소를 입력하면 됨.

#### **그외**

MX는 

```bash
> set type=MX
> n-study.com

```

```bash

> set type=NS
> n-study.com

<br>

### **DNS를 이용한 서버 부하분산**

대량 요청 처리로 인해 서버 여러대를 운용하고 있음.

#### **DNS 라운드 로빈 방식의 통신 flow**

>`'DNS서버'`와 `'부하를 분산하기 위한 서버군'`뿐인 간단한 방식.

예시:

www.example.com 웹사이트로 들어오는 요청을 서버1, 서버2, 서버3의 웹 서버 3대로 부하분산함.

1. PC1이 DNS 서버에게 www.example.com의 DNS쿼리를 송신.
2. DNS서버는 DNS응답으로 www.example.com의 IP주소 192.168.1.1을 반환.
3. PC1은 서버1(192.168.1.1)로 접근.
4. PC2가 DNS서버에게 www.example.com의 DNS 쿼리를 송신.
5. DNS 서버는 DNS응답으로 www.example.com의 IP주소 192.168.1.2를 반환.
6. PC2는 서버2(192.168.1.2)로 접근.

특별한 장비 필요X. 
- 요청에 순서대로 A레코드의 IP주소를 반환할 뿐.


! 효율적 분산은 불가능.

! 서버 장애도 감지 불가능. 계속 장애 발생한 서버 IP를 반환할 수도.

--> 이런 문제를 해결하기 위해 `'로드 밸런서'`라는 전용 부하분산 장비 존재.

<br>

#### **동적 DNS(DDNS)**

지금까지 배운 것: DNS가 NW상의 호스트명과 IP주소를 매핑해준다.

**! 매핑 정보는 변한다**

--> 호스트명과 IP주소 모두 설정으로 변경가능 + PC의 IP주소가 **DHCP에 의해 자동으로 설정**되는 경우(물론, 보통은 쓰던거 위주로 반환해줌)

매번 내가 갱신해줘야하나?

--> `동적 DNS`sys가 IP주소 변경시 DNS서버에 알려서 **동적으로 갱신**해줌.

예시:

1. 사내 NW의 동적 DNS는 DHCP에서 기능을 지원해서, 

    --> DHCP 서버가 PC에 IP할당할때마다 DNS서버에 리소스 레코드를 등록해줌.

2. 브로드밴드 라우터의 동적 DNS는 홈서버를 열었을 때, 집 밖에서 원격 접속을 하려면,

   브로드밴드 라우터의 WAN포트에 할당된 공인 IP주소로 통신해야함.

   **! 하지만**, 공인 IP를 일일히 기억하기 힘듦. 변경될수도.

   --> MS soft같은 곳에서 제공하는 DDNS 서비스 이용!

<br>

<br>

---


## **| TCP와 UDP**

**포트 번호 확인**

> PC에서 동작하는 app는 **포트번호**로 식별된다.
>
> 포트 번호는 자동할당됨.
>
> 이를 cmd에서 확인.

- `netstat` 명령어.

- 포트 번호를 질의하는 명령어구문: `netstat -n`

### **포트번호**

PC에는 여러 app이 실행중임(+백그라운드앱SV-데몬-까지) <=> 서버는 1가지역할만을 수행. (예: 메일 서버, 웹 서버 등등) 

실제 NW상에서 데이터 송수신하는 주체는 app(웹 브라우저 등).

==> 이런 app이 동작하는 pc나 서버는 IP주소로 특정할 수 있다.

==> **but**, app을 식별하는 정보도 필요했음.

==> **포트번호**등장.

> **pc나 서버에서 동작하는 app 그 자체를 식별하기 위한 정보.**
>
> - IP주소로 전달된 데이터는 다시, 포트 번호로 지정된 app에 전달된다.
> - 포트번호는 16비트.
>     - 0~1023 잘 알려진 포트
>     - 1024~49151 등록된 포트
>     - 49152~65535 동적/개인적 포트

- 등록된 포트: 자주 이용되는 app용 포트. 웹 프록시(proxy: 대리로 내부 nw나 인터넷 접속을 수행)서버의 8080

- 동적/개인 포트: client app을 식별하기 위한 포트. 서버에서 응답을 클라이언트로 해줄 때 사용.(통신은 양방향이어야 하니까! 응답포트가 필요하지.)

|프로토콜|tcp 포트번호|udp 포트번호|
|---|---|---|
|http|80|-|
|https|443|-|
|DNS|53|53|
|DHCP|-|67/68|
|FTP|20/21|-|
|SMTP|25|-|

**예를 들어,**


app사이의 통신은 연속적임. (이메일 송수신, 웹사이트 접속 등 여러 데이터가 연속해서 전송됨.)

==> 이런 app으로 주고받는 일련의 데이터를 `애플리케이션 플로우` 또는 `플로우`라고 부름.

이런 플로우는 '목적지/송신지 주소'와 '포트번호'로 식별된다.
- `IP주소:포트번호`

**<== 웹 사이트를 여러 개 띄울 수 있는 이유.**

1. 서버 app은 잘 알려진 포트 번호에서 클라이언트 pc로부터 접속을 기다린다.
2. 클라이언트 app은 잘 알려진 포트 번호를 목적지 포트 번호로 지정, 서버 app에 접속한다.

<== 서버 app이 특별히 설정하지 않는 이상, 명시적으로 포트 번호를 직접 지정하는 경우는 거의 없다.

<== 클라이언트는 OS에서 랜덤으로 동적/개인적 포트 번호에서 할당해줌. 서버로 보낼때에 잘 알려진 포트 번호 사용이 일반적.

<br>

### **TCP**

> Transmision Control Protocol.
>
> app간의 데이터 전달에 사용되는 프로토콜.

- `데이터 전달`
- `데이터 분할`
    - TCP의 데이터 size 최대치를 Maximum Segment Size(MSS)라고 함.
    - 전송할 데이터 크기가 MSS를 넘으면, 데이터를 분할함.
    - 실제 파일을 분할할 때, TCP header를 추가함. => TCP segment가 됨.
    - 근데, 당연한게 상위 L에서 내려오고, 전송 L를 지나갈때는 해당 header(= TCP header)를 붙여줌.



**| TCP 헤더 포맷**

|타입|개요|
|---|---|
|송신지포트번호|..|
|목적지 포트 번호|..|
|시퀀스 번호|송신데이터 순서를 나타내는 32bit. TCP segment의 순서를 확인|
|ACK 번호|데이터 수신 확인, 그 다음 수신하기를 기대하는 시퀀스 번호를 통지하기 위한 번호.|
|데이터 오프셋|TCP 데이터의 시작 위치 - 헤더길이를 나타냄|
|예약|예약돼있음 = 사용하지 않음.|
|플래그|URG(긴급), ACK(확인응답용. 연결 맺을 때 최초 세그먼트를 제외-반드시1) PSH(수신데이터를 바로 app에 전달), RST(연결강제종료), SYN(연결 성립요구), FIN(연결 정상종료)|
|윈도우사이즈|수신하는 쪽이 한 번에 수신할 수 있는 데이터 크기를 송신하는 쪽에 통지|
|체크섬|TCP헤더와 데이터 부분의 에러체크에 이용.|
|긴급 포인터|URG flag가 1일때만 사용. 긴급 데이터 시작 위치를 나타냄.|

<br>

#### **TCP 연결의 성립과 종료**


##### **TCP 연결성립**

TCP는 app간에 신뢰성 있는 데이터 전송을 수행하기 위해, `3-way handshake`라는 기법을 이용.

- 말 그대로 app끼리 인사를 3번해서 연결을 성립함.

예시:

1. 호스트A가 호스트B에게 SYN(연결해도 될까요?)신호를 보냄.
2. 호스트B가 호스트A에게 SYN/ACK(네,이쪽에서도 연결할래요)신호를 보냄.
3. 호스트A가 호스트B에게 ACK(네)신호를 보냄.
4. 호스트B가 호스트A에게 실제 데이터를 보냄.
5. 호스트A가 호스트B에게 이상없이 데이터를 받았다는 신호를 보냄.




##### **TCP 연결종료**

통신 종료시에도 규칙적으로 인사가 오감. 

- segment 4개가 오고가며 연결이 종료된다.

예시:

1. 호스트A가 호스트B에게 FIN/ACK(연결을 종료해도 될까요?)하고 신호를 보냄.
   - *ack는 서버로부터(B) 마지막으로 받았던 패킷에 대한 응답이다.*
3. 호스트B가 호스트A에게 ACK(네)신호를 보냄.
   - *서버(B)는 아직 보낼 게 남았을 수도 있으므로, 일단 단순 ack만 보냄.*
5. 호스트B가 호스트A에게 FIN/ACK(저도 연결을 종료할게요?)하고 신호를 보냄.
6. 호스트A가 호스트B에게 ACK(네)하고 신호를 보냄.

<br>

#### **데이터 전송 시의 동작**

TCP에서는 연결을 성립시킨 뒤, app끼리 데이터를 전송함.
- 이때, TCP는 '윈도우 제어', '재전송 제어', '혼잡 제어'와 같은 제어 기능을 이용함.

##### **윈도우제어**

얼만큼의 TCP segment들을 한 번에 보낼 것인가?를 제어하는 sys.

TCP에서는 데이터 수신 시, 송신지에 확인 응답을 돌려보냄.

==> 사실 이러면 느리잖아?

==> `윈도우 제어`로 **여러 개를 한 번에 송신**하자.

- TCP header의 윈도우 사이즈 부분에서 통지함.(서로의 윈도우 사이즈를 알려준다.)
     - 16비트니까 (2^16)-1 만큼 가능.

예시:

1. 3way handskae로 서로의 윈도우 사이즈를 확인. (SYN, SYN/ACK를 주고 받는 중에 윈도우 사이즈를 넣어준다.)
2. 상대의 윈도우 사이즈가 될때까진 플로우를 후루룩 보내본다.
3. 확인 응답을 받고 다시 플로우를 후루룩.

**이때! 윈도우 사이즈는 고정이 아니다.**

==> **대화 중, 확인 응답에 새로운 윈도우 사이즈를 끼워넣을 수 있다.**

이런 윈도우 제어를 `슬라이딩 윈도우`방식이라 부름.
> 송신하는 쪽이 전송할 데이터 전체를 수신하는 쪽의 윈도우 사이즈에 맞추어 MSS만큼 옆으로 옮겨가며 전송하는 이미지.
> - 윈도우라는 틀의 크기는 `윈도우 사이즈`, 움직이는 크기가 `MSS`.

<br>


한편, 플로우 보내다가 세그먼트 순서가 바뀌어버리면?

==> **순서제어**: 헤더의 시퀀스 번호를 이용한다.

<br>

##### **재전송 제어**

유실된 데이터를 재전송하는 sys.

예시:

1. A에서 B로 데이터를 보냄. RTT가(TCP segment을 보내고,이에 대한 확인 응답이 되돌아오는 시간) RTO(restransmission time out)시간이 지났는데도, 오지 않으면 '유실'이라고 판단.
2. 확인 응답이 유실됐으므로, 재전송.
3. 확인 응답 올 때까지 반복.

<br>

##### **혼잡 제어**

TCP segment가 유실되는 원인의 대부분은 NW혼잡에 있음.

==> 재전송 해봤자, 혼잡이 그대로면 의미가 있나?

==> **혼잡 제어**방식에는 여러가지가 있음.

1. `슬로우 스타트 알고리즘`

> a. 맨 처음 송신하는 쪽에서 혼잡 윈도우(송신쪽에서 한 번에 보내는 데이터 양)를 1로 설정함
> - 한 번에 윈도우만큼 보내는 게 아니라, 1 segment만 보내는 것.
> b. 확인 응답을 받음.
> c. 혼잡 윈도우를 1씩 늘림.

이렇게 점차 늘리는 방식. 

혼잡 제어는 재전송 타임아웃이 발생했을 때의 TCP 재전송에도 사용된다.

<br>

**결국, 이런 동작을 위해선 TCP 헤더에 많은 정보를 포함.** 필요 없는 정보로 **오버헤드**가 발생하지 않도록..!

==> **UDP를 이용**

<br>

### **UDP**

#### **UDP개요**

> User Datagram Protocol

TCP처럼 포트 번호에 따라 적절한 app 데이터를 전달하기 위한 프로토콜.

- UDP는 데이터 전달외에는 아무것도 안함.
- TCP처럼 연결을 성사하거나 재전송, 혼잡 제어 등을 안함.
- 데이터 순서 정렬도 하지 X.

==> **어쨌든 전달은 함**

<br>

##### **UDP 헤더포멧**

| 송신지 포트 번호 | 목적지 포트 번호 |
|---|---|
| 데이터그램 길이 | 체크섬 |


- 총 8B크기이다. <-> TCP 표준은 20B크기.

<br>

##### **UDP의 용도**

- 실시간 데이터를 전송할 때
    - *VoIP(Voice over IP): NW상의 보이스를 패킷으로 전송. 음성품질은 곧 지연이다.*
     

- 여러 상대에게 같은 내용의 데이터를 전송할 때
    - *TCP는 1대1 연결을 맺기에 유니캐스트. UDP는 브로드/멀티 캐스트가 가능함.*
    


- 데이터 전송 시 신뢰성이 필요치 않은 소량의 데이터를 전송할 때
    - *DNS 이름 해석처럼 소량 데이터를 빈번히 주고 받을 때.*
    - *큰 사이즈는 여러개 패킷으로 나눠 주고받기에, 중간에 유실되면 큰일남.*

<br>

##### **TCP와 UDP비교**

|   |TCP|UDP|
|---|---:|:---|
|통신방식|연결형|비연결형|
|신뢰성|높다|낮다|
|오버헤드|크다|작다|
|주요기능|애플리케이션 간의|데이터전송|
| |연결의 성립/유지/종료, 재전송 제어, 혼잡 제어, 윈도우 제어| |
|용도|시간 제약 없는 신뢰성이 필요한 데이터전송|실시간/브로드 $\cdot$ 멀티 캐스트/빈번한 소량데이터전송|

<br>

### **포트 열기**

인터넷 쪽에서 홈 nw내부의 애플리케이션으로 직접 통신을 시도할 경우, 보안상 차단됨.

==> **하지만**, 포트를 여는 설정을 통해 인터넷->홈nw 특정 pc로 통신이 가능함.

**<== 브로드밴드 라우터의 관리 페이지에서 포트를 여는 설정을 해보기.**

> 포트를 개방함으로써, 외부 인터넷 쪽에서 pc를 원격으로 조작하는 원격 접속 통신을 수행함.
>
> - 보통 TCP 3389번이 기본적인 원격 접속포트래.

1. 공유기 관리 페이지에서, TCP 3389(예시) 프로토콜: 원격 접속 통신 패킷을 192.~ 에 전송할 수 있도록 설정을 추가
2. 저장 = 규칙 저장

<br>

#### **포트를 연다는 것**

보안기능으로, 인터넷상으로부터 홈nw를 방어.
> 그 중 하나가 `'방화벽'`

보통 브로드밴드 라우터는 응답 데이터만을 홈nw로 들여보냄.

(부정 엑세스를 막아줌.)

##### **포트 개방**

> 인터넷쪽에서 먼저 통신을 시작해 **홈nw안 특정 pc로 패킷을 보낼 수 있도록** 한 것.

port forwarding, port mapping, port transmission등등 같은 말.

==> nw내부 특정 pc? 이거 **NAT.**

실제로 IP주소를 변환하기 때문에, 일종의 NAT에 해당한다.

>목적지 IP를 브로드밴드 라우터의 공인 IP로 하고, 
> 라우터에서 내부의 특정 IP주소로 바꿔주는 것.

<br>

##### **app 통신의 종류**

> 크게 두 가지로 나뉨.
>
> 1. `클라이언트/서버 형`
> 2. `피어 투 피어 형`

`클라이언트/서버 형`

- 클라이언트에서 서버로 요청,
- 서버에서 요청 처리결과 돌려줌.


클라이언트가 보통 먼저 시작.

`피어 투 피어 형`

- 대등하게 통신.

<br>

웹 브라우저나 이메일sw는 전형적 클라이언트/서버 형 app.

==> 따라서, 홈nw의 클라이언트로부터 인터넷 상 서버로 요청을 보낼 경우, 방화벽에 막히지X.

**But**, peer to peer app는 외부(홈nw의 외부)에서 먼저 통신을 시작할수도 있음.

==> 방화벽에 막힐 수도 있음.

예: skype, online game등

<br>

### **UPnP**

브로드밴드 라우터 포트를 열어줌으로써, PtoP app을 문제없이 이용할 수 있게 됨.

모르는 사용자들에게 포트를 여는 설정은 어려워...


==> **UPnP**(Universal Plug and Play)등장!

> NW장비를 연결하는 것만으로 다양한 기능을 이용할 수 있게하는 **기술 사양.**
>
> 이 중에 포트 개방이 있다. 많은 app과 nw 장비가 UPnP지원.

1. UPnP를 지원하는 app은(홈nw 안), 자신이 이용하는 포트 번호를 자동으로 브로드밴드 라우터에 통지함.
2. UPnP를 지원하는 라우터는, 인터넷 쪽에서 데이터가 들어오면, 해당 app이 동작중인 기기에 자동으로 전송해줌.

<br>

<br>

---

## **| TCP/IP 설정 이해**

### **기본 게이트웨이 설정**

> `기본 게이트웨이` = 같은 내부nw 내부의 `라우터`

네트워크는 라우터에 의해 서로 연결돼있다. 

==> **다른 nw로 가는 입구**에 해당하는 장비 = **라우터**

- PC의 기본 게이트웨이는 **브로드밴드 라우터의 IP주소**로 돼있음.
    - PC가 인터넷에 데이터를 보내면, 기본 게이트웨이(라우터)로 전송. 다른nw로 가야하면, 라우터가 목적지nw로 적절히 **라우팅**함.

다른 nw로 데이터를 보낼 때, '기본 게이트웨이(라우터)'에 보낸다는 점이 중요.

**But**, 라우터는 여러 개의 IP주소를 가지고 있음.
- 가정용 브로드밴드의 경우, LAN/WAN 총 2개의 IP주소를 가짐.

==> 내부에서는 기본 게이트웨이 IP로 **LAN쪽 IP**를 골라야 함. 

==> 데이터 전송은 **같은 nw 내부의 전송이 기본**이기 때문임.

**라우터의 포트마다 다른 nw이다.**

<br>

---
#### **!참고**

네, 정확하게 이해하고 계십니다! 라우터의 가장 핵심적인 역할이 바로 서로 다른 네트워크를 연결하고 구분하는 것이기 때문에, 말씀하신 내용이 모두 맞습니다. 👍

조금 더 명확하게 정리해 드릴게요.

##### 라우터의 핵심 역할: 네트워크의 경계선
라우터(Router)는 단어 뜻 그대로 '경로를 지정해주는 장비'입니다. A 네트워크와 B 네트워크가 있을 때, 둘 사이의 데이터가 오갈 수 있도록 길을 찾아주는 '네트워크 간의 교통경찰' 또는 **'환승역'**이라고 생각하시면 완벽합니다.

따라서 라우터의 각 포트(인터페이스)는 서로 다른 네트워크에 속해 있는 것이 기본 원칙입니다.

##### 질문에 대한 상세 답변
1. 라우터와 라우터 사이는 같은 네트워크인가요?
> 네, 맞습니다. 라우터와 라우터가 연결되는 그 '사이' 공간은 하나의 독립된 네트워크를 형성합니다.

예를 들어, 우리 집 공유기(라우터1)와 통신사의 장비(라우터2) 사이의 관계를 보면,

우리 집 공유기의 WAN 포트

통신사 라우터의 특정 포트

이 두 포트는 '통신사 하위 네트워크'라는 동일한 네트워크에 속해 있습니다. 그래서 서로 통신이 가능한 것입니다.

2. 포트 방향(LAN/WAN)으로 구분할 수 있나요?
> 네, 가정용 브로드밴드 라우터(공유기)에서는 LAN/WAN 포트가 바로 그 네트워크 경계를 명확하게 보여주는 부분입니다.

`LAN (Local Area Network) 포트 + Wi-Fi`

이들은 모두 **'하나의 내부 네트워크(NW)'**로 묶여 있습니다.

정확히는 공유기 내부에 있는 **'스위치'**라는 기능에 연결되어 있어, 이 포트들에 연결된 PC, 스마트폰, 게임기 등은 모두 같은 네트워크(예: 192.168.0.X)에 속하게 됩니다.

그래서 같은 LAN에 연결된 기기끼리는 라우터의 복잡한 경로 설정을 거치지 않고도 서로 빠르고 쉽게 통신할 수 있습니다. (파일 공유 등)

`WAN (Wide Area Network) 포트`

이 포트는 **'별개의 외부 네트워크(NW)'**와 연결되는 유일한 통로입니다.

이 포트를 통해 통신사(ISP)가 제공하는 네트워크에 연결되고, 이를 거쳐 인터넷이라는 거대한 세상으로 나아가게 됩니다.

3. WAN 쪽 네트워크의 경계는 또 다른 라우터까지인가요?
> 네, 이 또한 정확합니다. 우리 집 공유기의 WAN 포트가 연결된 네트워크의 경계는 바로 그 '다음 라우터', 즉 통신사의 라우터(또는 L3 스위치 등 관련 장비)까지입니다.

네트워크는 이렇게 라우터와 라우터를 거치며 계속해서 다른 네트워크로 연결되고 확장되는 구조로 이루어져 있습니다.

---

<br>

#### **기본 게이트의 다중화(VRRP)**

> Virtual Router Redundancy Protocol

라우터가 다운됐을 때, **다른 라우터가 기본 게이트웨이의 역할을 이어받음.**

- 라우터 여러 대를 통틀어 `가상 라우터`로 만듦.
    - 가상 라우터에는 IP주소가 있음.


- 가상 라우터 안의 물리적인 라우터에는 `마스터 라우터`와 `백업 라우터`로 나눌 수 있음.

1. 가상 라우터 IP가 기본 게이트웨이로 설정됨.
2. 가상 라우터로 전송된 데이터는 마스터 라우터가 처리함.
3. 마스터 라우터가 다운됐을 때, 백업 라우터들이 처리함.

<br>

### **DHCP**

> TCP/IP 설정 자동화 프로토콜

대부분 브로드밴드 라우터는 DHCP 서버 기능을 내장함.

자동으로 TCP/IP 설정을 수행.

- `ipconfig /release` 로 DHCP에서 할당된 ip주소를 해제.


```bash

무선 LAN 어댑터 Wi-Fi:

   연결별 DNS 접미사. . . . :
   링크-로컬 IPv6 주소 . . . . : fe80::9321:8650:6404:fa77%15
   기본 게이트웨이 . . . . . . :

```

위와 같이 없어졌다.

- `ipconfig /renew` 로 DHCP에서 ip주소 등 설정을 받아온다.

```bash
무선 LAN 어댑터 Wi-Fi:

   연결별 DNS 접미사. . . . :
   링크-로컬 IPv6 주소 . . . . : fe80::9321:8650:6404:fa77%15
   IPv4 주소 . . . . . . . . . : 172.17.98.99
   서브넷 마스크 . . . . . . . : 255.255.224.0
   기본 게이트웨이 . . . . . . : 172.17.96.1
```

#### **통신에 필요한 설정**

PC와 스위치를 LAN 케이블로 연결한 것만으로는 통신을 할 수 없음.

통신을 하기 위해서는 **IP주소를 비롯, TCP/IP 설정**을 해주어야 함.

#### **DHCP 시스템**

> DHCP를 이용하기 위해선 **DHCP 서버가 필요함.**

- 보통 windows werver나 Linux의 서버 OS에서는 DHCP의 기능이 표준으로 내장됨.
- 브로드밴드 라우터에도 DHCP기능 내장됨.

- DHCP서버는 DHCP 클라이언트에게 자동으로 할당할 IP주소의 범위를 설정해야함.
    - 설정할 **범위**를 **주소 풀**이라고 함.

> 주소 풀에는 
>- IP주소,
>- 서브넷 마스크,
>- 기본 게이트웨이 IP주소,
>- DNS 서버 IP주소
>
>도 자동으로 할당하도록 설정가능.

- DHCP에서 할당하는 Ip주소는 리스형식을 취하고 있다.

- IP주소를 할당받을 클라이언트는, DHCP 클라이언트 설정을 해야 함.

##### **IP 주소 할당 Sys.**

> 기본적으로 DHCP 클라이언트와 서버는 다음의 4단계로 작동함.

a. `DHCP DISCOVER`

b. `DHCP OFFER`

c. `DHCP REQUEST`

d. `DHCP ACK`

1. 먼저, **DHCP 클라이언트 A가** DHCP 서버를 찾아내기 위해 DHCP DISCOVER를 BroadCast함. (IP주소를 모른다 가정)
   
3. DHCP DISCOVER를 받은 **서버가** DHCP OFFER를 돌려줌. (할당 가능한 IP주소와 기본 게이트웨이 등 DHCP 옵션이 포함됨.)

4. **클라이언트가** DHCP OFFER를 받으면, DHCP REQUEST를 BroadCast해서 실제 IP주소를 요청함.
    - 브로드캐스트의 이유는, 한 개의 subnet상에서, DHCP 여러 대가 존재할 가능성 고려.
    - 따라서, 어떤 DHCP 서버의 OFFER를 선택한건지 포함된 REQUEST를 보냄.

5. 선택된 DHCP 서버를 제외한 **나머지 서버에선**, 해당 REQUEST한 IP주소를 사용 불가능으로 설정함.

6. DHCP REQUEST를 받은 **서버는** DHCP ACK를 보냄으로써 클라이언트가 Ip주소 등의 TCP/IP 설정 정보를 이용할 것을 허용함.

7. DHCP ACK를 받은 **클라이언트는** 수신한 IP주소가 이미 이용되고 있진 않은지 확인 후(ARP로 브로드캐스트해보고), 설정 정보를 적용함.

8. **서버는** 할당한 IP주소와 클라이언트의 MAC 주소의 매핑을 저장함.

9. 필요에 따라 **클라이언트는** DHCP REQUEST를 송신해 임대기간을 연장함.

10. 이제 IP가 필요없어지면(종료 등), DHCP RELEASE를 서버에 전송해 해제함.
    - 수동으로 했던게 `ipconfig /release`
    - 다시 설정정보를 요청하는게 `ipconfig /renew`

<br>

`DHCP 릴레이 에이전트`는, 

**DHCP 서버 1대**가 **여러 nw상의 클라이언트들**에게 IP주소 등의 설정 정보를 **할당**해주는 기능.

> 브로드캐스트된 DHCP DISCOVER를,
>
> 유니캐스트로 변환해,
>
> DHCP 서버에 전송하는 기능.이라고 보면 됨.

1. 클라이언트 A가 DHCP DISCOVER를브로드캐스트.
   
3. 라우터에서 DHCP 릴레이 에이전트 기능으로, UNICAST로 바꾸어 서버에 전송.
   
5. DHCP 릴레이 에이전트의 IP주소를 보고 DHCP 풀 1의 IP주소를 배포한다.
    - 미리 서버에 DHCP 풀을 여러 개 설정해 두어야 함.

- DHCP DISCOVER에는, DHCP 릴레이 에이전트의 IP주소가 들어 있음.

- DHCP 서버는 DHCP 릴레이 에이전트의 IP주소로부터, 할당해야 할 DHCP 풀을 판단해서, DHCP 클라이언트에 적절한 설정 정보를 보냄.

### **웹 사이트 접속**

#### **웹사이트를 본다(Web Access)는 의미**

> 웹 서버로부터,
>

이때, 

- 웹사이트 내용을 기술하기 위해 사용하는 언어가 HTML(Hyper Text Markup Language)

- 웹 서버로부터 파일을 올바르게 전송받기위한 프로토콜이 HTTP(HyperText Transfer Protocol)

#### **웹 엑세스 흐름**

1. user가 URL을 웹브라우저에 입력.

2. URL로 지정한 웹 서버에 HTTP 요청을 송신.

3. 웹 서버가 요청을 해석.

4. 웹 서버가 요청받은 파일을 웹 브라우저에게 돌려보냄.

5. 웹 브라우저가 수신한 데이터를 해석해 표시.

실제론 웹 서버는 여러 대. 

웹 페이지 하나에도 여러 서버에서 통신하고 있다.

- 요청하지 않은, 광고 서버 등에 의해 통신이 발생함.

#### **URL**

> 웹 사이트 주소에 해당. Uniform Resource Locator.

인터넷 상의 데이터위치를 가리키는 표준 표기법.

| http:// | www.naver.com | /network/index.html |
|---|---|---|
| 스킴 :// | 호스트명 | /path |

- `스킴scheme` : web browser가 서버의 데이터에 접근할 때 프로토콜.
      - : 뒤는 데이터의 위치, //는 바로뒤에 호스트명이 온다는 뜻
  
- `호스트명` : 포트번호는 생략(바로 뒤에 오게돼있음)

- `path` : 목적 데이터가 웹 서버 내 어디에 있는지.

#### **정적 웹페이지/ 동적 웹페이지**

- `동적 웹페이지` : 웹 브라우저의 요청에 따라 웹 서버가 자동으로 생성한 웹 페이지. ( 검색결과 페이지, 장바구니 페이지 등....)

#### **HTTP**

> 웹 페이지는 HTML로 쓰여진 HTML파일과 HTML 파일 내에서 불러오는 이미지 파일 등으로 구성.
>
> 웹 서버에 존재하는 HTML/이미지 파일을 PC의 **웹 브라우저로 전송하기 위해 이용되는 프로토콜.**


HTTP 요청과 HTTP 응답을 주고받는다.

##### **HTTP요청**

 웹브라우저가 웹 서버로 보내는 HTTP요청은
 - `요청 행` Request Line
 - `메시지 헤더` Message Header
 - `엔티티 바디` Entity Body

 의 3부분으로 나눔.

<br>

**요청 행**은, HTTP 요청의 첫 번째 라인에 위치. 웹 서버에게 처리 요구를 전달.

요청 행은 `메소드`, `URI`, `버전`으로 구성됨.
- 메소드: 서버에 대한 요구 커맨드
    - GET : URI로 지정한 데이터를 가져옴.
    - HEAD : URI로 지정한 데이터의 응답 헤더만을 가져옴.
    - POST : URI로 지정한 서버의 커맨드에 대해 데이터를 송신함.
    - PUT : URI로 지정한 데이터를 전송한 데이터로 덮어씀.


      

- URI: 요청 대상인 데이터를 가리킴.
- 버전: 웹 브라우저가 지원하는 HTTP의 버전.

<br>

**메시지 헤더**는, 요청 행 다음에 옴. 웹 브라우저 종류나 버전, 데이터 형식 등의 정보.

**엔티티 바디**는, 파일을 보낼 때 사용. 보낼 데이터 없으면 안 씀.

<br>

HTTP예시:

```http
GET/network/index.html HTTP/1/1

User-Agent:Mozilla/4.0(~)
...
...
Accept -~
Accept -~
Refer.http://www/n- ~
Cache- ~
--------------
-----------


```

#### **HTTP 응답**

> 마찬가지로 `상태 행`, `메시지 헤더`, `엔티티 바디`의 3부분으로 나뉨.

**상태 행**은 `버전`, `상태코드`, `설명문`으로 나뉨.

상태코드는 웹 서버의 처리 결과를 세 자리 숫자로 나타냄.

|상태코드|의미|
|---|---|
|1xx|정보, 추가 정보가 있다는 걸 알림|
|2xx|성공, 서버가 요청을 성공적 처리함|
|3xx|리다이렉트, 다른 URI로 다시 요청을 보낼 것|
|4xx|클라이언트 에러, 요청에 문제가 있음|
|5xx|서버 에러, 서버 쪽에 문제가 있음|

#### **HTTP 쿠키**

> 웹 서버가 웹 브라우저에게 특정한 정보를 저장해두기 위한 sys.

웹 서버는 쿠키로 유저의 로그인 정보나 사이트 내 웹 페이지의 열람 이력 등을 관리.

- 관련 상품을 띄우는 등의 웹페이지 커스터마이즈 가능
- HTTP는 원칙적으로 stateless(1번 연결에 1번만 전송가능). 쿠키를 이용해, 접속 상태를 보존 (stateful).
    - 다음 접속부터 HTTP요청시 서버에 쿠키도 같이 보내는 것.

### **메일**

#### **이메일 시스템**

> 이메일 송수신에는 `SMTP`와 `POP3`라는 2개의 프로토콜을 조합해 사용함.
>
> - SMTP는 메일 송신을 위한 프로토콜
>
> - POP3는 메일 서버로부터 자기 앞으로 온 메일을 수신하기 위한 프로토콜.

초기에는 p2p였지만, 목적지 호스트가 전원이 꺼져있거나, 하는 경우를 대비해 메일서버 방식을 사용함.

메일 송수신을 위해서 메일 sw를 사용.
- PC의 TCP/IP 설정
- 메일 계정 설정 필요

1. 이메일을 송신하면 메일 데이터는 SMTP에 의해 사전에 설정된 메일 서버로 보내짐

2. 수신한 메일 서버는 받는 사람의 메일 주소 도메인명으로부터 목적지 메일 서버를 판단함.

3. SMTP로 목적지 메일 서버에 메일을 전송함.

4. 목적지 메일 서버에서는 받는 사람의 메일주소ID를 보고 각 유저의 편지함에 메일을 저장함.

5. 메일을 수신할 유저는 자기 메일 서버의 편지함에 메일이 와있는지 확인가능.

6. 와 있으면 POP3로 자기 컴퓨터에 메일을 다운로드 후 읽음.


<br>

##### **이메일 주소**

> `ID @ 도메인명` 형식으로 표시됨.

1. 이메일sw는 받는/보내는 사람의 이메일 주소를 지정해서 이메일을 보냄.

2. 송신한 메일은 받는 사람의 메일 주소 **도메인명**에 해당하는 메일 서버까지 SMTP로 전송됨.
- 도메인명으로 IP주소를 찾는, DNS를 사용함. MX레코드에서 목적지 메일 서버의 호스트명을 알고, 다시 한 번 IP주소를 알아냄.

### **스위치의 구조**

스위치의 역할
> - PC나 서버 등을 nw에 연결시킨다.
> - 같은 nw 내부의 LAN 포트 사이에서 데이터 전송을 수행한다.

스위치에는 LAN 포트가 여러 개 달려있음.

스위치를 여러 대 사용하는 경우에도 같은 NW로 취급됨.

**스위치는 어느 LAN포트로 보낼지 'MAC주소'로 확인함.**

<br>

#### **MAC 주소**

> LAN 포트의 주소( HW주소 / 물리적 주소 - 변경을 웬만해선 못해)

- 48bit
  
- 8bit씩 8번 끊어읽음
  
- 상위24비트는 `벤더 코드` (장비업체) 예: 시스코는 00-00-0C
  
- 하위 24비트는 NW장비 벤더가 제조한 LAN 포트를 관리하기 위한 '시리얼번호'로 사용됨.

- 모두 1인 MAC주소를 `브로드캐스트`주소라고 한다.

<br>

#### **스위치의 데이터 전송**

> 스위치는 MAC주소로 데이터를 전송함.

스위치는 `MAC주소테이블`을 가지고 있음.
- 전원이 막 켜진 경우, 텅 비어있음.
  

다음과 같은 방식으로 MAC주소테이블이 채워짐.



1. MAC주소 A에서  목적지 MAC 주소 B로 송신한다.

2. 스위치에 패킷이 도착, 송신지 주소를 보고 MAC주소 테이블에 등록함.
    - 예를 들어, 포트 1번은 MAC주소 A라는 식.



3. 목적지 MAC주소와 MAC 주소테이블로 목적지를 판단.

4. 이 경우, 테이블이 비어있으므로, 포트 1이외의( 들어온 포트 이외 ) 모든 포트로 전송. = **Flooding**함.

5. MAC주소가 아직 등록되어있지 않을때에도 Flooding을 함. = Broadcast
    - Broadcast주소는 FF-FF-FF-FF-FF-FF.
    - Broadcast가 닿는 범위를 `브로드캐스트 도메인`이라함.
    - 브로드캐스트 도메인이란, `NW 하나의 범위`

이번에는 B에서 A로 보낸다 하면, B가 들어온 포트와 B를 테이블에 작성하고, A는 이미 있으니 A에 매핑되는 포트로 보냄.

MAC 주소 테이블은, MAC의 포트 연결이 언제든 바뀔 수 있으므로, 시간 제한이 있다.

단, PC가 연결돼 데이터를 전송하는 동안에는 등록이 유지.

<br>

#### **ARP**

> IP주소마냥 직접 MAC주소를 지정하기란 어렵다.
>
> 외우기도 쉽지 않고
>
> ==> **ARP**(Address Resolution Protocol) 등장!

**IP 주소에 매핑된 MAC 주소를 알아내기 위한 프로토콜이다.**

- 통신할 때에는 반드시 IP주소를 지정해야한다.

- IP 주소에 매핑된 MAC주소를 알아내기 위해 ARP를 사용한다.

**대략적인 ARP 요청**

1. ARP 요청은 브로드캐스트로 송신됨.

2. 같은 NW 내부의 모든 PC를 대상으로, 지정된 IP 주소에 매핑된 MAC 주소를 질의.

3. IP주소에 해당하는 PC가 ARP 응답으로 MAC주소를 알림.

**구체적인 ARP 요청 예시**

스위치 포트 1-4에, 각각 MAC 주소 A-D가 연결돼 있다.

1. IP주소 192.168.1.4( = D)를 지정해 통신을 시작한다.

2. 지정된 IP주소의 MAC 주소를 알아내기 위해서 A가 자동으로 ARP요청을 송신한다.

3. 스위치에서, ARP요청은 브로드캐스트로 플러딩한다.

4. 질의 된 IP주소를 가진 D가 ARP 응답을 반환한다.

5. ARP로 MAC주소를 알아냈으니, 이제 실제 데이터를 전송한다.

단, 위에는, 목적지가 같은 nw상에 존재할 경우.

**다를 경우**에는, ARP 요청의 대상이 되는 IP주소는 **기본 게이트웨이의 IP주소**.

$\because$ **Boradcast는 같은 nw안에서만 되니까** (Flooding). 또, ARP 대상인 IP주소는 같은 NW안의 IP주소로 한정된다.

---
! 참고

아니요, **ARP를 직접 시행하는 주체는 스위치가 아닙니다.**

ARP의 주체는 IP 주소를 가지고 통신해야 하는 **단말 장치(PC, 서버 등)나 라우터**입니다. 스위치는 그 과정에서 중요한 역할을 하지만, 단지 '우편배달부'처럼 메시지를 전달만 해줄 뿐입니다.

#### ## 역할 구분: 질문자와 우편배달부 🤔

이 과정을 아파트에서 사람을 찾는 것에 비유하면 이해하기 쉽습니다.

* **ARP 주체 (PC, 라우터) = '질문자'**
    "101동 502호(IP 주소)에 사는 분, 얼굴(MAC 주소) 좀 알려주세요!"라고 질문을 시작하고, 답변을 받아 기록하는 역할입니다.

* **스위치 = '아파트 복도' 또는 '메가폰' 📢**
    502호를 찾는다는 외침(ARP 요청)을 **복도에 있는 모든 집**에 들리도록 전달해주는 역할을 합니다. 스위치는 그 외침의 내용("누가 502호인가?")을 이해하지 못합니다. 단지 모든 곳으로 소리를 퍼뜨릴 뿐입니다.

#### ## 각 장비의 정확한 역할

1.  **PC (또는 라우터)**: **(ARP 시행 주체)**
    * 통신하려는 상대방의 IP 주소는 알지만 MAC 주소를 모를 때, 자신의 **ARP 캐시** 테이블을 먼저 확인합니다.
    * 정보가 없으면 "이 IP 주소(예: 192.168.1.5)를 가진 사람은 누구인가?"라는 내용의 **ARP 요청(Request) 메시지**를 만듭니다.
    * 이 요청을 **브로드캐스트(Broadcast)** 방식으로 네트워크에 전송합니다.

2.  **스위치**: **(단순 전달자)**
    * PC로부터 ARP 요청 프레임을 받습니다. 이 프레임의 목적지 MAC 주소는 `FF:FF:FF:FF:FF:FF`(브로드캐스트)입니다.
    * 스위치는 L2 장비이므로 이 메시지의 IP 주소 내용은 확인하지 않습니다.
    * 단지 목적지가 브로드캐스트인 것을 보고, 프레임이 들어온 포트를 제외한 **연결된 모든 포트로 이 메시지를 그대로 복사하여 전달(Flooding)**합니다.

3.  **목적지 PC**: **(응답자)**
    * 스위치가 전달해 준 ARP 요청을 받고, 메시지 안의 IP 주소가 자신의 것임을 확인합니다.
    * "그 IP 주소는 내가 사용하고 있으며, 내 MAC 주소는 이것이다"라는 내용의 **ARP 응답(Reply) 메시지**를 만들어 **유니캐스트(Unicast)** 방식으로 최초 질문자에게 직접 보냅니다.



#### ## 예외: L3 스위치

물론, **L3 스위치**처럼 라우팅 기능이 있는 스위치는 IP 주소를 이해하고 **라우터처럼 동작**합니다. 따라서 VLAN 간 통신 등을 위해 자신이 직접 ARP 요청을 보내거나 응답할 수 있습니다. 하지만 이는 스위치의 기본 기능(L2)이 아닌 라우팅 기능(L3)을 수행할 때의 이야기입니다.

#### !참고


자세한 설명
1. ARP는 동일 네트워크(L2)에서만 동작합니다.
**ARP(Address Resolution Protocol)**의 근본적인 목적은 **'같은 네트워크 구간'**에 있는 장비의 IP 주소(L3)를 물리적인 MAC 주소(L2)로 변환하는 것입니다.

통신의 기본 원리: 컴퓨터가 실제로 데이터를 전기 신호로 바꿔서 케이블로 보내려면, 최종적으로는 상대방의 MAC 주소를 알아야 합니다. IP 주소는 전체 인터넷에서 사용하는 논리적인 주소일 뿐, 실제 데이터 전송(프레임)은 MAC 주소를 기반으로 이루어집니다.

ARP의 한계: ARP는 브로드캐스트를 사용해서 "이 IP 주소 가진 사람 누구야? MAC 주소 좀 알려줘!"라고 외칩니다. 이 외침은 같은 네트워크, 즉 브로드캐스트 도메인 안에서만 유효합니다.

2. 브로드캐스트는 라우터를 넘어가지 못합니다.
이것이 가장 핵심적인 이유입니다.

브로드캐스트(Broadcast): 네트워크에 연결된 모든 장비에게 데이터를 한 번에 전송하는 방식입니다. ARP 요청이 바로 이 방식을 사용합니다.

라우터(Router)의 역할: 라우터는 서로 다른 네트워크를 연결해주는 장비입니다. 라우터의 중요한 기능 중 하나는 브로드캐스트를 차단하는 것입니다. 만약 라우터가 브로드캐스트를 모든 곳으로 전달한다면, 전 세계 인터넷은 불필요한 브로드캐스트 트래픽으로 마비될 것입니다. 이를 '브로드캐스트 스톰(Broadcast Storm)'이라고 합니다.

이 두 가지가 결합되어 나타나는 동작
목적지 판단: 내 PC(예: 192.168.0.10)가 구글 서버(예: 8.8.8.8)로 데이터를 보내려고 합니다.

네트워크 비교: PC는 자신의 서브넷 마스크(예: 255.255.255.0)를 이용해 목적지 IP 주소(8.8.8.8)가 자신과 다른 네트워크에 있다는 것을 인지합니다.

게이트웨이 필요성 인지: PC는 '아, 이건 내가 직접 보낼 수 없으니 외부로 나가는 문, 즉 기본 게이트웨이(라우터, 예: 192.168.0.1)에게 보내야겠다'라고 판단합니다.

게이트웨이의 MAC 주소 찾기: 이제 PC의 목표는 구글 서버가 아닌 기본 게이트웨이에게 데이터를 전달하는 것으로 바뀝니다. 데이터를 전달하려면 기본 게이트웨이의 MAC 주소가 필요합니다.

ARP 요청 발송: PC는 같은 네트워크 안에 있는 기본 게이트웨이(192.168.0.1)의 MAC 주소를 알아내기 위해 ARP 요청을 브로드캐스트합니다.

응답 및 데이터 전송: 기본 게이트웨이가 자신의 MAC 주소를 알려주면, PC는 아래와 같이 이더넷 프레임을 만들어 전송합니다.

목적지 MAC: 기본 게이트웨이의 MAC 주소

출발지 MAC: 내 PC의 MAC 주소

목적지 IP: 구글 서버의 IP 주소 (8.8.8.8)

출발지 IP: 내 PC의 IP 주소 (192.168.0.10)

이 프레임을 받은 기본 게이트웨이는 IP 주소를 보고 "아, 이건 구글 서버로 가는 거구나"라고 인지한 뒤, 다음 경로로 패킷을 전달(라우팅)해 줍니다.

!참고

네, 아주 좋은 질문입니다. `route-server.ip.att.net` 같은 인터넷 라우트 서버는 일반적인 라우터와는 역할이 조금 다릅니다.

간단히 말해, 인터넷 라우트 서버는 **실제 데이터를 전달(포워딩)하지 않고, 인터넷 세상의 '경로 지도(Route Map)'를 모아서 보여주는 '공개 전망대'** 같은 역할을 합니다.


#### 인터넷 라우트 서버(Route Server)란? 🗺️

인터넷은 수많은 네트워크(AS, Autonomous System)들의 연결로 이루어져 있습니다. 각 네트워크는 BGP(Border Gateway Protocol)라는 약속을 통해 "우리 네트워크로 오려면 이 길을 이용해!"라고 서로의 경로 정보를 교환합니다.

**라우트 서버**는 이 BGP 경로 정보를 한곳에 모아, 네트워크 관리자나 일반 사용자가 **'특정 네트워크(여기서는 AT&T)의 관점에서 인터넷 경로가 어떻게 보이는지'**를 들여다볼 수 있게 해주는 특별한 서버입니다.

가장 중요한 특징은 다음과 같습니다.

* **데이터를 전달하지 않습니다**: 일반 라우터처럼 여러분의 유튜브 영상이나 웹사이트 데이터를 직접 전달하는 역할은 하지 않습니다. 오직 경로 정보(BGP 라우팅 정보)만 교환하고 보여주는 역할만 합니다.
* **'루킹 글래스(Looking Glass)' 역할**: 이름처럼 인터넷의 내부를 들여다보는 '유리창' 역할을 합니다. 이를 통해 외부에서 특정 네트워크의 라우팅 테이블을 확인할 수 있습니다.



#### ## 라우트 서버가 정확히 하는 일 🔍

사용자께서 `route-server.ip.att.net`에 접속해서 `show route` 명령어로 보신 것은 바로 AT&T의 BGP 라우팅 테이블입니다. 이 서버는 다음과 같은 중요한 역할을 합니다.

##### 1. **BGP 경로 정보 수집 및 제공**

전 세계 수많은 ISP와 기업들이 이 라우트 서버에 자신의 BGP 경로 정보를 보냅니다. 라우트 서버는 이 정보들을 모두 취합하여 거대한 '인터넷 경로 지도'를 만듭니다. 사용자는 이 서버에 접속해 특정 목적지까지 어떤 경로들이 있는지 확인할 수 있습니다.

**예시:** `show route 8.8.8.8` 명령어를 치면, AT&T의 네트워크에서 구글 DNS(8.8.8.8)까지 가는 경로가 어떤 네트워크(AS 번호)들을 거쳐 가는지 그 목록(AS_PATH)을 전부 볼 수 있습니다.

##### 2. **네트워크 문제 해결 및 분석** 👨‍💻

네트워크 엔지니어들은 라우트 서버를 통해 다음과 같은 작업을 합니다.

* **라우팅 정책 검증**: 자신들이 광고한 네트워크 경로가 다른 ISP(AT&T)에게 제대로 전달되고 있는지 확인할 수 있습니다.
* **연결 문제 진단**: 특정 웹사이트나 서비스에 접속이 안 될 때, 라우트 서버에 접속해 경로 정보가 아예 없는지, 아니면 비정상적인 경로로 설정되어 있는지 파악할 수 있습니다.
* **최적 경로 분석**: 여러 경로 중 어떤 경로가 가장 효율적인지(짧은지) 분석하여 자사 네트워크의 라우팅 정책을 최적화하는 데 사용합니다.

##### 3. **투명성 및 연구 자료 제공**

주요 ISP들이 라우트 서버를 공개적으로 운영함으로써 인터넷의 경로 정보가 어떻게 구성되고 전파되는지에 대한 투명성을 높입니다. 또한, 연구자들은 이 데이터를 분석하여 인터넷 전체의 안정성이나 구조에 대한 연구를 진행하기도 합니다.



결론적으로, `route-server.ip.att.net`에서 하신 경험은 **AT&T라는 거대 통신사의 관제탑에 앉아 전 세계로 뻗어 나가는 항공기들의 항로 지도를 직접 열람**한 것과 같습니다. 실제 승객(데이터)을 태우고 운항하지는 않지만, 모든 항로 정보를 보고 분석할 수 있는 매우 유용한 도구인 셈이죠.

네, 맞습니다. **대부분의 경우 라우터의 '인터페이스(Interface)'는 물리적인 '포트(Port)'를 의미**하는 것이 맞습니다. 랜선을 꽂는 이더넷 포트(Ethernet Port)나 광케이블을 연결하는 SFP 포트 등이 바로 물리 인터페이스입니다.

하지만 정확히 말하면, **'인터페이스'가 '포트'보다 좀 더 넓고 포괄적인 개념**입니다.



#### ## 포트 vs. 인터페이스: 무엇이 다른가?

모든 **물리적 포트**는 **인터페이스**이지만, 모든 **인터페이스**가 **물리적 포트**인 것은 아닙니다. 인터페이스는 크게 두 종류로 나눌 수 있습니다.

##### 1. 물리적 인터페이스 (Physical Interface)

이것이 바로 우리가 흔히 말하는 **'포트'**입니다.

* **정의**: 장비에 실제로 존재하여 케이블을 연결할 수 있는 하드웨어 연결 지점입니다.
* **역할**: 데이터가 라우터를 드나드는 실제 입구와 출구 역할을 합니다.
* **예시**:
    * **이더넷 인터페이스 (Ethernet Interface)**: `GigabitEthernet0/0`, `FastEthernet0/1` 등으로 표시됩니다.
    * **SFP 인터페이스**: 광케이블 모듈을 장착하는 포트입니다.



##### 2. 논리적 인터페이스 (Logical/Virtual Interface)

물리적인 형태 없이 **소프트웨어적으로 생성된 가상의 인터페이스**입니다. 특정 기능이나 목적을 위해 만들어지며, 눈에 보이지 않습니다.

* **정의**: 물리적 연결 없이 라우팅, 관리 등의 목적으로 소프트웨어상에서 만들어진 가상 통로입니다.
* **역할**: 라우팅 프로토콜 관리, VLAN 간 통신, 가상 사설망(VPN) 연결 등 다양한 역할을 수행합니다.
* **예시**:
    * **루프백 인터페이스 (Loopback Interface)**: 라우터 자체를 가리키는 가상 인터페이스입니다. 물리 포트와 달리 항상 활성화 상태이므로 라우터의 고유 주소를 할당하거나 테스트, 관리 목적으로 사용됩니다.
    * **VLAN 인터페이스 (SVI)**: VLAN(가상 근거리 통신망) 간의 라우팅(통신)을 위해 생성하는 가상 인터페이스입니다.
    * **터널 인터페이스 (Tunnel Interface)**: VPN과 같이, 네트워크 위에 또 다른 가상 통로를 만들어 데이터를 전송할 때 사용됩니다.


#### ## 간단한 비유 🏢

* **라우터** = 하나의 **건물**
* **물리적 인터페이스 (포트)** = 건물의 **출입문, 창문** (외부와 실제로 연결되는 통로)
* **논리적 인터페이스** = 건물 **내부의 각 방을 연결하는 문** (외부와 직접 연결되진 않지만, 건물 내에서 공간을 나누고 이동하는 통로)

#### ## 핵심 요약

| 구분 | **포트 (Port)** | **인터페이스 (Interface)** |
| :--- | :--- | :--- |
| **개념** | **물리적인** 연결 단자 | 데이터가 드나드는 **논리적인** 통로 |
| **범위** | 인터페이스의 한 종류 (하위 개념) | 물리적 포트와 논리적(가상) 인터페이스를 모두 포함 (상위 개념) |
| **종류** | 이더넷 포트, SFP 포트 등 | 물리 인터페이스, 루프백, VLAN, 터널 인터페이스 등 |

결론적으로, "라우터에 인터페이스가 여러 개 있다"고 할 때는 **물리적인 포트뿐만 아니라, 필요에 따라 생성된 가상의 통로들까지 모두 포함**해서 이야기하는 것입니다.

   - 라우팅 프로토콜

2. `nw ID/subnet mask`

   라우터는 데이터의 목적지 IP주소를 키로 사용해서, 라우트 정보의 nwID/subnet mask를 검색함.

3. `메트릭`

   **라우터로부터 목적 nw까지의 거리를 수치화 한 것.**

   물리적인 거리가 아님.

   라우팅 프로토콜에 의해 학습된 라우트 정보에 포함돼 있음.

   가장 작은 루트 = **최적 루트**

4. `관리 거리`(AD, Administrative Distance)

   라우팅 프로토콜 마다 계측하는 목적 nw까지 거리.

   프로토콜마다 다른 메트릭을 비교할 수 있도록 조절하는 파라미터가 **'관리 거리'**

   즉, **라우터는 AD와 metric에 의해 목적 nw까지 거리를 인식한다.**

5. `다음 홉 주소`(Next hop address)

   'hop'은 다음 라우터를 가리킴.

   다음 전송해야할 라우터의 IP주소를 말함.

   원칙적으로 라우터와 같은 nw안에 있는 다른 라우터의 IP주소.

6. `출력 IF`

   목적 nw로 패킷을 전송할 때, 패킷을 출력할 IF의 정보.

   **다음 홉 주소와 출력 IF를 합친 것이 목적 NW로 가는 방향**이다.

7. `경과 시간`

   라우팅 프로토콜로 학습한 라우트 정보마다, 라우팅 테이블에 등록된 시점으로부터 경과한 시간이 기재.

   경과 시간이 길면 길수록 안정된 라우트 정보.

   

<br>

이런 **정보 등록의 방법 3가지**

1. `직접 연결`

   라우터가 직접 연결된 nw의 라우트 정보.

   - 라우터의 IF에 IP주소를 설정하고, 그 IF를 활성화만 시키면 자동으로 라우팅 테이블에 등록됨.

2. `정적 라우팅/ 라우팅 프로토콜`

   앞에서 말했듯, 라우터는 **라우팅 테이블에 등록된 nw에 한해서만 IP 패킷을 라우팅 가능.**

   = 직접 연결된 nw밖에 모름.  $\because$ 직접 연결은 자동으로 등록하거든.

   $\therefore$ 원격 nw는 라우팅 테이블에 등록해주어야 함.

   ==> 이 방법이 **정적 라우팅과 라우팅 프로토콜**

- '정적 라우팅'은 유저가 **수동**으로,
- '라우팅 프로토콜'은 **자동**으로. = 라우터끼리 정보를 교환한다.
    - `RIP` : 규모 작은 소기업 nw
    - `OSPF` : 중~대기업 nw
    - `BGP` : 앞의 실습처럼, 주로 인터넷 상의 라우터에 이용됨.

#### \#\# 1. 디폴트 라우트와 기본 게이트웨이

**네, 맞습니다.** 라우팅 테이블에 나오는 \*\*디폴트 라우트(Default Route)가 가리키는 곳이 바로 기본 게이트웨이(Default Gateway)\*\*입니다.

  * **디폴트 라우트**: PC의 라우팅 테이블(경로 지도)에 등록된 '특별 규칙'입니다. 보통 목적지가 `0.0.0.0`으로 표시되며, "이 지도에 명시되지 않은 **모든 외부 목적지**는 무조건 여기로 보내라"는 의미입니다.
  * **기본 게이트웨이**: 그 '여기'에 해당하는 실제 주소, 즉 외부 네트워크로 나가는 관문 역할을 하는 라우터의 IP 주소입니다.

PC는 인터넷으로 나가려고 할 때 자신의 라우팅 테이블을 보고, 목적지(예: https://www.google.com/search?q=google.com)로 가는 경로가 없으면 디폴트 라우트 규칙에 따라 모든 패킷을 기본 게이트웨이로 보냅니다.



#### \#\# 2. 네트워크 주소와 게이트웨이 주소

**이것도 정확하게 보셨습니다.** 일반적으로 네트워크 주소는 마지막 자리가 `.0`이고, 기본 게이트웨이는 `.0`이 아닌 다른 숫자를 사용합니다.

##### **네트워크 주소 (Network Address)** 🗺️

  * **네트워크 자체를 나타내는 대표 주소**입니다. `192.168.0.0`은 '192.168.0.x' 대역을 사용하는 네트워크 전체를 의미하는 '거리 이름'과 같습니다.
  * **보통 끝이 `.0`인 이유**: 가장 널리 쓰이는 서브넷 마스크인 `255.255.255.0` (/24)을 사용할 경우, 네트워크 주소를 계산하면 마지막 자리가 항상 0이 되기 때문입니다. 이는 규칙이라기보다는 가장 일반적인 설정의 결과입니다.
  * 이 주소는 네트워크 전체를 상징하므로 **개별 PC나 장치에 할당할 수 없습니다.**

##### **기본 게이트웨이 주소 (Gateway Address)** 🚪

  * 네트워크의 '출입문' 역할을 하는 라우터에 할당된 **실제 IP 주소**입니다.
  * **끝자리가 `.0`이 아닐 수 있는 이유**: 게이트웨이도 해당 네트워크에 속한 하나의 장치(호스트)이므로, **해당 네트워크에서 사용 가능한 IP 주소 중 하나**를 할당받아야 합니다.
  * `192.168.0.0/24` 네트워크를 예로 들면, IP 주소는 `192.168.0.1`부터 `192.168.0.254`까지 할당할 수 있습니다. (네트워크 주소인 `.0`과 브로드캐스트 주소인 `.255`는 제외)
  * 관습적으로 관리하기 쉽게 \*\*가장 첫 번째 주소인 `.1`\*\*이나 가장 마지막 주소인 `.254`를 게이트웨이 주소로 사용하는 경우가 많지만, **규칙은 아니므로 `192.168.0.100` 같은 다른 주소를 사용해도 전혀 문제없습니다.**

#### \#\# 핵심 정리

  * **디폴트 라우트 (`0.0.0.0`)**: "모르는 길은 전부 이쪽으로\!"라는 **규칙**입니다.
  * **기본 게이트웨이 (예: `192.168.0.1`)**: 위 규칙이 가리키는 실제 **라우터의 주소**입니다.
  * **네트워크 주소 (예: `192.168.0.0`)**: 네트워크 전체를 나타내는 **대표 주소**로, 장치에 할당할 수 없습니다.
  * 결론적으로, 사용자의 PC는 인터넷에 접속하기 위해 `0.0.0.0` 규칙에 따라 모든 외부행 패킷을 `192.168.0.1` 주소를 가진 라우터로 보내고 있는 것이 맞습니다.

##### **기업 nw의 방화벽**

> 똑같이 필요한 통신 데이터만을 허가.



기업 nw의 방화벽은 **DMZ**(DeMillitarized Zone)이라는 nw를 구성함.

인터넷에서 직접 접근이 가능한 **공개서버**를 운용하기 때문에, 이런 공개서버를 설치하기 위한 nw인 DMZ를 운용함.

기업용 방화벽은 반드시 3개 이상.

`사내 nw`, `DMZ`, `인터넷 쪽의 nw`를 상호 연결함.

- 인터넷으로부터 => DMZ에 설치된 공개 서버로 오는 통신은 통과 O.

- 인터넷으로부터 => 사내 nw로 오는 통신은 통과 X.

<br>

##### **DoS를 막는 IPS**

DMZ nw위의 공개서버는 방어가 안돼... => 방화벽이 완벽하진 않음.

$\therefore$ **DoS**(Denial of Service)같은 공격취약.

$\therefore$ **IPS**(Intrusion Prevention Sys.)같은 시스템이 도입됨.

`IPS`는 방화벽과 같이 nw상 전송되는 **데이터를 체크함.**

단시간에 요청이 반복되거나, 비정상적인 요청을 포함한 데이터 등, **징후를 검출, 차단.**

##### **암호화 개요**

nw 데이터는 도청위험, 변조 위험 방지를 위해 `데이터 암호화`

- 암호화 되지 않은 데이터를 `평문`이라고 함. 

> **암호화**란, **암호 키**를 이용해 평문을 암호문으로 바꾸는 것.

암호문을 만드는 sys를 가리켜 '암호화 알고리즘'이라고 부름.

`암호화 알고리즘`에 따라 암호키로 평문 데이터에 **어떤 연산처리**를 하는지 정해짐.

#### **암호화 방식**

##### **대칭 키 방식**

> 암호화와 복호화에 같은 암호 키를 사용하는 방식.
> - 암호 키를 안전하게 배송할 필요가 있음.
> - `DE3/3DES`, `AES`가 있음.

##### **공개 키 암호 방식**

> 암호화와 복호화에 다른 암호 키를 이용하는 방식.
>
> 비대칭 키 방식이라고도 한다.
>
> **수신자**가 **키 페어**를 만들어 **공개 키**는 모두에게 공개, **비밀 키**는 본인이 가짐.
>
> - 공개 키로 페어인 키를 알아내는 게 불가능은 아님.
> - `RSA 암호`, `타원곡선 암호` 등이있음.
> - 대칭키에 비해 암호/복호화 부하가 크다. 

<br>

#### **디지털 서명과 인증서**

##### **디지털 서명**

공개 키 암호 방식을 사용할 때,

데이터 변조 여부와 데이터 송신자를 확인하기 위해 사용하는 게 **디지털 서명**

**송신자가 데이터에 디지털 서명을 추가함.**

> `디지털 서명`이란, 데이터로 계산한 `해시값`을 송신자의 비밀 키로 암호화한 것.
>
> - 해시 값 게산에는 `MD5`, `SHA-1`등 알고리즘 사용.
> - 입력하는 데이터가 달라지면 = 변조되면, 해시 값도 달라짐.

데이터를 수신한 수신자는,

1. 데이터 부분으로부터 해시 값을 계산.
2. 송신자의 공개 키로 디지털 서명을 복호화.
3. 수신자가 계산한 해시값과, 복호한 해시 값 비교.
4. 또한, 송신자의 공개 키로 디지털 서명을 복호화할 수 있다면, 송신자는 페어인 비밀 키를 가지고 잇는 올바른 송신자라는 것 확인.

**$\therefore$** 공개 키로 비밀키로 암호화한 걸, 복호화 할 수 있다. 이걸 이용해서, **if 디지털 서명을 복호화 == 데이터 복호화**를 확인함.

##### **인증기관**(CA)

인증서는 인증기관이라는 제 3자 기관에서 발행함.

인증서 안에 공개키가 들어 있음.

- `GeoTrust`
- `CyberTrust`
- `Versign` 등등

인증기관끼리는 서로를 신뢰, `'인증서 체인'`을 구성함.

인증 기관이 발행한 증명서를 이용한 sys 기반을 `PKI`(public key infrastructure)이라고 함.



**인증서를 발급받기 위한 과정**

1. 키 페어를 만든다.
2. 공개 키와 소유자 정보를 인증 기관에 보내 인증서 발급 신청을 한다.
3. 인증서 발급 심사
4. 인증서를 발급

<br>

### **SSL**

**PKI**, 다시 말해 인증 기관이 발행한 인증서의 이용 사례로 가장 많은 것이 **SSL**(Secure Socket Layer)에 의한 암호통신.

- https://로 시작하는 URL을 가진 웹사이트의 통신은 모두 SSL로 암호화된 것.

SSL은 인증서를 통해 정규 서버(위장X서버)와의 사이에서 암호화된 안전한 통신 제공.

**SSL 통신**은 **공개 키 암호 방식 & 대칭 키 암호 방식을 조합**해 사용.

1. 통신을 시작할 때, 먼저 통신 상대의 인증서를 체크(공개 키 암호 방식으로)해 위장여부확인.
2. 데이터 전송 시에는, 대칭 키 암호 방식을 사용한다.(고속 처리)

<br>
