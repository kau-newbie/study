# **| 클라우드 컴퓨팅이란? |**

> 정보 처리를 인터넷 너머 <span style="background-color:#FAFAB4">다른 사람의 컴퓨터 자원으로</span> 처리하는 서비스.

컴퓨터와 소프트웨어를 자신이 소유하지 않고, **<span style="color:#3C5087">네트워크를 통해, 클라우드 사업자의 컴퓨터와 소프트웨어를 사용함.</span>**

(클라우드(cloud) = 인터넷의 생김새를 비유함.)

<br>

<span style="color:gray">참고 | NIST(미국 국립 표준 기술연구소)의 정의: " 클라우드 컴퓨팅이란, 공유 구성이 가능한 컴퓨팅 리소스(NW, Server, storage, app. sv)의 통합을 통해 어디서나 간편하게, 요청에 따라 NW를 통해 접근하는 것을 가능하게 하는 모델이다. 이는 최소한의 이용절차 또는 서비스 공급자의 상호 작용을 통해, 신속히 할당되어 제공된다."</span>

<br>

#### NIST의 클라우드 특징 5가지 

1. **주문형 셀프 서비스**

   
    - 사업자의 수동 조작 없이, 사용자별 관리 화면에서 서비스를 이용할 수 있다.

2. **광범위한 네트워크 접속**

    - 모바일 기기 등 다양한 디바이스로 서비스에 접속할 수 있다.

3. **리소스의 공유**

    - 여러 사용자가 같은 컴퓨팅 리소스를 공유하여 이용한다. 사용자마다 리소스가 할당되지만, 사용자는 시스템의 어느 부분에 접속했는지 알 수 없다.
      

4. **신속한 확장성**

    - 필요에 따라, 필요한 만큼 컴퓨팅 리소스를 늘리거나 줄일 수 있다.

5. **측정 가능한 서비스**

    - 서비스를 이용한 만큼 요금이 부과된다.

<br>

<br>

### | 클라우드가 등장한 배경 |

1. **메인 프레임(1980~)**

    - 대형 범용 컴퓨터 '메인 프레임'의 시대. 모든 데이터와 애플리케이션을 메인 프레임이 모아서 처리함.
    - 단말기는 입/출력 용도로만 사용함.
     


2. **분산형 클라이언트 서버모델(1990~)**


    - 클라이언트 단말기에 처리기능을 탑재해서, 분산 처리로 변화함.
      



4. **네트워크 컴퓨팅(2000~)**

    - 웹 브라우저를 이용한 애플리케이션이 회사 인트라넷에 구축. 처리가 다시 서버에 집중됨.
      



5. **클라우드 컴퓨팅(2010~)**

 
     - 서버를 보유하는 것이 아니라, 전 세계에 분산된 서버의 리소스들을 서비스로써 이용하는 모델임.
      

<br>

이같은 발전 과정은,


**1. 기술 발전 측면**


    - CPU 처리 속도 고속화.
    - 가상화기술과 분산처리기술 발전.
    - 빠르고 저렴한 네트워크.
    - 거대해진 데이터 센터에 의한 '규모의 경제' 등의 이유가 있음.




**2. 기업 사용자 측면**

   
    - IT 투자 비용의 절감, 유연한 서비스 설계이용, 구축 및 운용 부담의 경감 등이 과제가 됨.

        - 경제성, 유연성(리소스의 확장/축소), 가용성(하드웨어의 장애에도 유연한 대처가능),  빠른 구축 속도 등등....


**3. 클라우드 사업자 측면**

    컴퓨팅 자원을 셀프서비스(사용자가 직접 필요한 만큼 선택함) 형태로 제공하면,
    
    - 서비스 제공의 효율이 높음.
    - 지속적인 매출(안정적 매출)에 유리.

## **클라우드 컴퓨팅 서비스 구분**

<br>

서비스 제공 형태가 다양해져 구분하기 어려움. 하지만 보통 크게 3가지로 나눔.

1. SaaS (SW as a SV)
2. PaaS (Platform as a SV)
3. IaaS (Infrastructure as a SV)

<br>

<br>

### **SaaS (서비스로서의 소프트웨어)**

> 주로 업무에서 사용하는 SW 기능을, <span style="background-color:#FAFAB4">인터넷 등의 네트워크를 통해, 필요한 만큼 SV로 이용할 수 있도록</span> 제공하는 형태.

기존의 패키지 SW처럼 모든 기능을 라이센스별로 구매X

--> **<span style="color:#3C5087">필요한 기능(SW)을 필요한 기간만큼 임대</span>** 하는 방식.

<br>

- **하나의 서버를 여러 기업에서 공유**하는 것을 전제하는, **멀티 테넌트 방식** 서비스.
  - 단, 데이터는 사용자별로 분리되도록 설계 -> 보안성 확보

- SW 업데이트 작업은 클라우드 사업자가 수행.

- SV 계약 후 사용자 계정만 있으면 바로 SV를 이용할 수 있음.
    - 패키지 SW 구입시와 비교하면, 도입부터 납기까지 간격이 매우 짧음.

- 인터넷을 통해 활용하므로, **다양한 디바이스로 접속할 수 있음**.

<br>


SaaS로 제공하는 `SV`:  
- MS Office 365, 구글의 G Suite 등.

SaaS로 제공하는 `SW`: 
- 정보시스템의 전자메일
- 그룹웨어 <span style="color: gray">(그룹웨어는 기업이나 조직 내에서 구성원 간의 원활한 소통과 협업을 지원하는 소프트웨어. 이메일, 전자결재, 캘린더, 메신저, 문서관리 등 다양한 기능을 하나의 플랫폼에서 제공하여 업무 효율성을 높이는 데 목적이 있음.)</span>
- CRM(Customer Relationship Management, 고객 관리 시스템) 등.




<br>

#### **| 사업자의 운용 관리 범위**

| --- | `SaaS` | `PaaS` | `IaaS` |
|---|---|---|---|
|애플리케이션|o|---| --- |
| 미들웨어 | o | o | --- |
| OS | o | o | --- |
| HW | o | o | o |

[참고: 미들웨어](https://aws.amazon.com/ko/what-is/middleware/ "미들웨어란?")

<br>

<br>

<br>

### **PaaS (서비스로서의 플랫폼)**

> 기업의 애플리케이션 <span style="background-color:#FAFAB4">실행 환경 및 애플리케이션 개발 환경을 서비스로써 제공</span>하는 모델.

<br>

자사에서 애플리케이션 개발 환경을 처음부터 구축하는 건 시간이 많이 듦.

- --> Java, PHP, Ruby 등의 언어를 지원하는 애플리케이션 실행 환경이나, 데이터베이스 등이 미리 마련되어 있음. 
- --> **<span style="color:#3C5087">인프라 구축 및 운용 보수를 하지 않아도, 제공된 기반을 사용 가능함.</span>**
- --> 빠르게 응용 프로그램을 개발, 서비스를 제공할 수 있음.

<br>

대신, <span style="color:#D2691E">서버 및 미들웨어의 상세한 설정은 할 수 없음. </span>


또, 특정 PaaS 환경에 의존도가 높아져 <span style="color:#D2691E">다른 환경으로 전환(migration)이 어려움.</span>

<br>

주된 용도는, 

- 개발 및 테스트 시행에 큰 처리 능력이 필요한 경우.

- 자사에서 운용 중인 애플리케이션의 최대 부하를 분산 처리하는 경우.

- 스마트폰 등 모바일 서비스에서 인터넷 접속이 필요한 경우, IoT 센서 데이터를 수집하는 경우 등등.

PaaS로 제공하는 `SV` 및 `SW`:
- 사이보우즈의 kintone
- 오픈소스 SW인 Cloud Foundry, OpenShift 등.

<br>

**| 주요 PaaS 제공 업체:**





1. `AWS` (Amazon Web Services):
    > 광범위한 클라우드 서비스를 제공하며, PaaS 솔루션으로 Elastic Beanstalk, AWS Lambda 등을 포함합니다.




2. `Google Cloud`:
    > Google Kubernetes Engine (GKE) 및 App Engine과 같은 PaaS 서비스를 제공합니다.



3. `Microsoft Azure`:
    > Azure App Service, Azure Kubernetes Service (AKS) 등 PaaS 솔루션을 제공합니다.




4. `Red Hat`:
    > 오픈소스 기반의 PaaS 솔루션인 OpenShift를 제공합니다.



5. `Salesforce`의 Heroku:
    > 웹 애플리케이션 개발 및 배포에 특화된 PaaS 플랫폼을 제공합니다. 


<br>

**| 국내 PaaS 회사:**


- `K-PaaS`:
    > 국내 클라우드 기술력을 기반으로 한 플랫폼으로, 민간 주도 생태계 구축을 목표로 합니다. 
- `국내 클라우드 서비스 제공사` (CSP):
    > KT, 네이버, NHN 등도 자체 PaaS 솔루션을 제공하며 시장 경쟁에 참여하고 있습니다. 

<br>

#### **| PaaS로 제공되는 대표적인 도구와 서비스**

|대분류|소분류|상세|
|---|:---:|---|
| 개발도구, 부속 서비스 | API SV | 인증, 과금, 알림, 분석 등의 부가 서비스 |
|                     | SDK | 모바일용 SW 개발도구 등 |
|   | 개발 FW | Ruby on Rails, Sinatra, Node.js, Eclipse 등 |
| 핵심 기능 | 프로그래밍 언어  | Ruby, Java, Python, PHP 등 |
|   | App 서버 | Apache, Tomcat 등 |
|   | DB 서버 | MySQL, MongoDB, Amazon RDS, Oracle DB 및 MS SQL 서버 등|
|   | 메시징 미들웨어 | RabbitMQ, Amazon SQS 등 |
|   | 다른 SV의 지원 | 추가 기능, API와의 연계 등 |
|   | 기타 | App 통합, 비즈니스 프로세스 관리, 데이터 통합, <br> 관리파일 전송, 포털, 보안, 테스트 환경 등|

### **IaaS (서비스로서의 인프라 스트럭쳐)**

> CPU나 HW 등의 <span style="background-color:#FAFAB4">컴퓨팅 리소스를 NW를 통해 SV로 제공</span>하는 모델.

<br>

IaaS의 사용자는 자사 단말기에서 메뉴를 선택하면, 서버와 스토리지, 방화벽 기능과 로드 밸런싱 기능 등을 이용할 수 있음.

<span style="color:gray">참고: 로드 밸런싱이란, "IT 시스템에서 트래픽이나 작업 부하를 여러 서버에 고르게 분산시켜주는 기술. <br>쉽게 말하면, 한 서버에 너무 많은 요청이 몰리지 않도록 균형을 맞춰주는 역할."</span>


<br>

IaaS로 제공하는 `SV`:

- 가상 서버 및 온라인 스토리지 등

> `가상 서버`란, 
>
> 클라우드 사업자가 보유하는 물리적 서버의 CPU, 메모리, 스토리지 등의 **<span style="color:#3C5087">HW자원을 SW적으로 나누어 사용자에게 제공</span>** 하는 것을 말함.

--> 기업 사용자는 물리적 서버를 구입하지 않고, 필요할 때 필요한 만큼 가상 서버를 만들 수 있다.

<span style="color:#D2691E">단, 가상 서버에서는 OS에 설치되는 DB와 미들웨어, APP 등의 SW를 직접 설치하고 관리해야 함.</span>

활용 예시:

> 웹 사이트 서버 - 프로모션 이벤트용(사용자가 단기간에 몰림) - 를 운용할 때, 해당 기간동안만 리소스를 많이 빌림.

<br>

- 보통 종량제나 월 정액제로 판매되고 있음.

- 데이터 전송량에 따라 추가요금 부과하는 방식도 있음.

<br>

**| IaaS 대표 SV**

- `Amazon Web Services`의 Amazon Elalsti Compute Cloud(`EC2`) 등

<br>

<br>

<br>

`PaaS`와 `SaaS`의 차이점은 정해진 SW를 SV받는게 아닌, <span style="background-color:#FAFAB4">자사에서 개발한 애플리케이션을 가동한다는 점.</span>

 `PaaS`와 `IaaS`의 차이점은 <span style="background-color:#FAFAB4">서버, NW, 보안 부분을 클라우드 사업자에게 위임한다는 점.</span>

## **클라우드 컴퓨팅 이용 모델 구분**

<br>

클라우드 환경을 누가, 어디에 구축하고 운영하느냐에 따라 부르는 이름이 달라진다.

    1. 퍼블릭 클라우드
    
    2. 프라이빗 클라우드
    
    3. 커뮤니티 클라우드
    
    4. 하이브리드 클라드

<br>

### **퍼블릭 클라우드**

> 클라우드 사업자가 sys를 구축, 인터넷망 등의 <span style="color:#3C5087">**NW를 통해 불특정 대상에게 SV를 제공**</span>하는 형태.

<span style="color:#3C5087">클라우드 sys이</span> 기업 또는 개인의 **<span style="color:#3C5087">방화벽 외부에 구축</span>**

--> 사용자는 이 모델에서 자사의 IT 자산을 보유하지 않더라도 컴퓨팅 리소스를 SV로 사용.

--> 필요한 컴퓨팅 자원을 단기간에 저비용 마련, 운용 관리 부담이 적음.

<br>

<span style="color:#D2691E">단, 클라우드 사업자의 정책에 따라야 함.</span>

<br>

**| 가상 프라이빗 클라우드(VPC)**


퍼블릭 클라우드의 자원을 활용해, 특정 기업 사용자 전용 클라우드 환경을 설계하는 형태도 존재. 

 --> vpn망 등, 폐쇄망 접속을 통한 전용 환경제공.

<span style="color:gray">copilot</span> 

> <span style="color:gray">" VPC는 마치 공용 건물 안에 있는 전용 사무실 같아요.
건물은 퍼블릭 클라우드, 사무실은 VPC.<br>
문을 잠그고, 출입증을 설정하고, CCTV를 설치하면
인터넷 기반이지만 완전히 통제된 공간이 되는 거죠."</span>

<br>

### **프라이빗 클라우드**

> 사용자의 데이터 센터에, 클라우드 기술을 응용한 <span style="color:#3C5087">**자사 전용 환경을 구축**</span>하여, 컴퓨팅 리소스를 유연하게 이용할 수 있는 형태.

--> 가상화, 자동화같은 클라우드 관련 기술의 활용으로 인해 sys의 성능과 비용이 최적화.

<br>

자사의 NW에 직원용으로 구축하는 환경. 자사 정책에 맞추어 구축할 수 있음.

<span style="color:#D2691E">단, 높은 기술력과 운용력과 비용이 필요하다.</span>

<br>

#### **프라이빗 클라우드 종류**

 최근 두 가지로 분류됨.

1. `온프레미스 프라이빗 클라우드`
 
2. `호스티드 프라이빗 클라우드`

**1. 온프레미스 프라이빗 클라우드**

기업 사용자가 장비를 자사의 데이터 센터 등에 설치하고, 자사가 보유하고 운용하는 형태.


>자체적으로 구입한 HW(서버 및 스토리지, 그 외 리소스)를 설치.
>
> 이렇게 구축한 클라우드 기반 SW 등을 스스로 이용.


<span style="background-color:#FAFAB4">대규모일 때, 자사 전용 클라우드 환경을 구축해서 운용하는게 유리함.</span>

- 자사의 sys 요구 사항에 맞춘 유연한 sys 설계가 가능함. 

- 자체적인 보안 정책에 따른 강력한 보안 환경을 구축해서 운용함.

<br>

**2. 호스티드 프라이빗 클라우드**

클라우드 사업자가 장비를 보유, 프라이빗 클라우드 sys을 제공하는 형태.

> 전용선이나 VPN망을 통해 보안 수준이 높은 전용 클라우드 환경을 제공받음.
> 
> 고객 전용 HW로 따로 분리해놓음. <-> VPC는 논리적으로 분리된 망.

<br>

<span style="background-color:#FAFAB4">온프레미스(On-Premises) sys?</span>
> 기업 사용자가 자체 설비로 운용하는 서버, 네트워크, 저장장치 등의 IT 환경.

<br>

<br>

`온프레미스 sys`의 경우
- 도입시에 데이터 센터 등의 시설 비용과 서버, 스토리지, NW장비 등의 **HW & SW 조달비용**이 필요.
- **이용량 예측**이 필요. 
- 도입 후, HW 리스 $\cdot$ 임대 비용, 유지 비용, 회선 비용 등.



> 단, <span style="color:#3C5087">**5년 이상의 장기간에 걸쳐 sys를 사용하는 경우, 혹은 대규모 sys를 구축하고 운용할 경우**</span> 온프레미스가 저렴한 경우도 있음.
>
> 또, 클라우드 구축시, 기존 온프레미스sys에서 migration 시 비용이 부담스러움.

`클라우드 sys`의 경우

- 계약 기간을(리스나 임대) 걱정하지 않고, 필요에 따라 자원을 유용하게 사용가능함.
- **운영 비용 평준화**됨.

### **클라우드의 리스크**

<br>

클라우드 sys 자체적인 <span style="color:#D2691E">리스크</span>들이 있음.

<br>

예를 들어, `IaaS`의 경우
> 클라우드 사업자의 **HW장애**
> 
> **NW리스크**로 sys다운위협, 통신도청, 중간자 공격, 스푸핑 같은 통신위협

--> 가상 서버를 백업하거나, 쉽게 가상 서버 환경을 구축하기 위한 템플릿을 마련해 대비.

<span style="color:gray">템플릿? : 가상서버 템플릿은 서버를 만들 때 필요한 요소들을 미리 설정해둔 구성 파일이나 이미지.</span>

<br>

<span style="color:#D2691E">다른 리스크</span> 예시 

- 갑작스런 클라우드 사업자의 SV 중단 (파산 등) 위험이 존재함. (클라우드 SV의 연속성 리스크)
- 데이터센터 HW들의 고장 리스크.
- 재해나 운영자의 조작 실수 등.


<br>

#### **클라우드 보안 거버넌스**



<span style="color:gray">교재: (Governance, 기업의 경영진이 클라우드를 이용할 때의 위험을 주체적이고 적절하게 관리하기 위한 구조를 구축하고 운용하는 것)</span>

# **| 클라우드 서비스와 그 이용법 |**

> 클라우드 SV로는 가상서버, 스토리지, NW, DB같은 다양한 기능이 있음.
>
> 대부분은 클라우드 사업자들이 제공함.

<br>

`가상 서버`는 물리적 서버와 CPU, 메모리, 스토리지 같은 HW 자원을 SW로 논리적으로 나누어 사용하는 것.

- 실제 서버와 마찬가지로, OS 및 App을 동작시킬 수 있음.

- 가상 CPU 성능과 메모리 용량을 선택 가능함.

클라우드 SV로 제공되는 **로드 밸런서**의 기능을 활용하면, 가상 서버를 이중화하거나 부하를 분산할 수 있음.

또, 통신량에 따라 가상서버의 갯수를 자동으로 조정함(오토 스케일).

<br>

`스토리지 SV`는 데이터, 콘텐츠 아카이브, 백업, 파일 서버의 사용, 시스템의 재해 대책(Disaster Recovery)등 다양한 용도로 사용됨.

<br>

`클라우드 접속시`

1. 인터넷을 이용. 

2. 높은 수준의 보안이 요구될 때:  VPN 망, 전용선 등의 NW SV를 이용할 수 있음.

<br>

`DB SV`는 기업의 기간계 (Mission Critical/ Core) sys, 빅데이터, IoT등, 이용 목적에 맞게 다양한 DB 존재함.

<br>

이러한 **서비스들을 선택하여 조합.**

--> 소규모~ 대규모 sys를 구성할 수 있음.

### 🏆 데이터베이스의 장점 (vs 스토리지)

| 항목 | 데이터베이스(DB) | 스토리지(Storage) |
|------|------------------|-------------------|
| 🔍 **검색 속도** | 인덱스, 쿼리 최적화로 빠른 검색 가능 | 파일 이름이나 경로 기반 검색만 가능 |
| 🧠 **데이터 구조화** | 테이블, 관계, 제약 조건 등으로 정형화 | 구조 없음, 단순 파일 저장 |
| 🔄 **데이터 처리 기능** | CRUD(생성, 조회, 수정, 삭제), 트랜잭션 처리 | 파일 업로드/다운로드만 가능 |
| 🔐 **보안 및 권한 관리** | 사용자별 접근 제어, 암호화, 감사 로그 | 파일 단위 권한 설정만 가능 |
| 🔄 **동시성 제어** | 여러 사용자의 동시 작업을 안전하게 처리 | 충돌 방지 기능 없음 |
| 🧩 **데이터 무결성** | 제약 조건으로 데이터 오류 방지 | 무결성 보장 어려움 |
| 📈 **확장성과 분석 기능** | 대량 데이터 처리, 통계, BI 연계 가능 | 분석 기능 없음, 별도 도구 필요 |


요약하자면, DB는 데이터를 “잘 다루기 위한 도구”고, 스토리지는 단순히 “보관하는 장소”야.

### 📚 데이터베이스의 역사 간단 정리

#### 1. **1950~60년대: 초기 개념 등장**
- 미국에서 군사 정보 관리 목적으로 등장
- "데이터의 기지"라는 의미로 **Database**라는 용어 사용

#### 2. **1970년: 관계형 데이터베이스(RDBMS) 탄생**
- IBM의 **에드거 F. 커드(Edgar F. Codd)**가 관계형 모델 제안
- SQL 언어 개발 → 데이터 검색과 조작이 쉬워짐

#### 3. **1980~90년대: 상용 DBMS 확산**
- Oracle, IBM DB2, Microsoft SQL Server 등 등장
- 기업의 ERP, 금융, 제조 시스템에 도입

#### 4. **2000년대: 웹과 빅데이터 시대**
- NoSQL 등장 (MongoDB, Cassandra 등) → 유연한 구조와 확장성
- 클라우드 기반 DB 서비스 (AWS RDS, Azure SQL 등) 확산

#### 5. **현재: 하이브리드와 AI 시대**
- NewSQL, 시계열 DB, 그래프 DB 등 다양한 형태
- AI/ML과 연계된 실시간 분석 DB도 등장

### 로드 밸런서

**로드 밸런서(Load Balancer)**는 서버에 몰리는 트래픽을 **여러 서버에 고르게 분산**시켜주는 장치 또는 소프트웨어야. 쉽게 말하면, **트래픽 분산 관리자**라고 보면 돼.



### 🚦 로드 밸런서란?

> 클라이언트 요청을 여러 서버에 **균등하게 분배**해서  
> **성능, 안정성, 확장성**을 높이는 기술 또는 시스템



#### 🧩 왜 필요한가?

- **트래픽 집중 방지**: 한 서버에 요청이 몰리면 다운될 수 있음
- **장애 대응**: 특정 서버가 죽어도 다른 서버로 자동 전환
- **확장성 확보**: 서버를 추가해도 자동으로 분산 처리
- **성능 향상**: 응답 속도 개선, 지연 시간 감소



#### ⚙️ 작동 방식

로드 밸런서는 클라이언트 요청을 받아서 다음 기준에 따라 서버에 전달해:

- **라운드 로빈**: 순서대로 분배
- **최소 연결 수**: 현재 연결이 가장 적은 서버로
- **가중치 기반**: 성능 좋은 서버에 더 많은 요청 할당
- **IP 해시**: 클라이언트 IP 기반으로 고정 서버 할당

### **가상 서버 SV**

> 클라우드 SV를 구성하는 가장 기본적인 요소.



1. <span style="background-color:#FAFAB4">개별 서버(물리 서버)가 보유한 CPU나 메모리 등을 논리적인 HW 리소스로 간주.</span>

2. 개별 서버가 보유한 리소스를 여러 개의 가상 서버가 나누어 사용.

3. 혹은, 다수의 개별 서버가 보유한 리소를 한 대의 가상 서버가 통합해서 사용.

단, 가상 서버는 HW를 `에뮬레이션`하기 때문에, 가상화하지 않은 서버에 비해 성능이 떨어지기 마련.

- 그래서 고성능 서버가 필요할 경우, 실제 서버(베어메탈 서버)를 제공하는 경우도 있다.

---

#### ! 참고

🖥️ 그럼 VMware는 에뮬레이션일까?
VMware는 "가상화(Virtualization)" 기술을 사용하는데,
에뮬레이션과는 비슷하지만 다소 다른 개념이야.
| 구분 | 에뮬레이션 | 가상화 (VMware 등) | 
|---|---|---|
| 목적 | 다른 시스템을 흉내냄 | 같은 시스템을 격리된 환경에서 실행 | 
| 성능 | 느릴 수 있음 (완전한 재현) | 거의 네이티브 수준 | 
| 예시 | 게임 에뮬레이터, CPU 에뮬레이터 | VMware, VirtualBox, Hyper-V | 
| 호환성 | 다른 아키텍처도 가능 (예: ARM → x86) | 보통 같은 아키텍처에서만 작동 | 


즉, VMware는 에뮬레이터가 아니라 하이퍼바이저 기반의 가상 머신을 제공하는 도구야.
하지만 넓은 의미에서 보면 "에뮬레이션의 한 형태"로 볼 수도 있어.


**VM Ware는 가상 머신이라면, BlueStack이나 NOX는 에뮬레이터.**

#### **가상 서버의 이용**

- 보통 셀프 서비스식 포털화면으로 조작. (기동/중지/재기동/삭제 등)

- 가상 서버의 SV 플랜은 가상 CPU의 성능 및 메모리 용량이 정해져있음.

- 가상 서버에는 시동 디스크로 root disk가 할당됨. 

- 그 외 OS 선택 등등..

1. 가상 서버를 생성할 리전(SV 제공지역)선택

3. 가상 서버에 접속하기 위한 엑세스 키 생성
 
5. 가상 서버에 연결을 허용하는 통신 유형 설정

7. 이용하는 OS와 루트 디스크를 선택(HDD나 SSD등)

9. 사용할 가상 서버 스펙 선택(CPU나 메모리)
 
11. 이용하는 가상 서버의 상세 설정(가상 서버를 기동하는 리전과 NW 설정 등)

13. 추가 디스크 선택(스토리지)

15. 설정한 내용으로 서버 만들고 시작

#### **가상 서버에서 사용할 수 있는 옵션기능**

> 대다수의 클라우드 사업자가 제공하는 대표적인 기능들이 있음.
> - 로드 밸런서
> - 오토 스케일
> - 정기적인 스냅샷

##### **로드밸런서**

가상 서버 사용 시 '리전'과 '존'을 선택할 수 있음.

- `리전(region)`이란, '서울, 도쿄 등 지리적으로 떨어진 독립적인 데이터 센터 묶음(지역)'을 뜻함.

- `존`이란, 같은 리전(지역) 내의 독립적인 데이터센터를 뜻함.

<br>

🧱 구조를 비유로 설명:

리전 = 도시 전체

존 = 도시 안의 서로 다른 건물들

서버 = 건물 안의 컴퓨터들

- 각 존은 독립된 건물처럼 전력, 냉각, 네트워크가 분리돼 있음
- 서버는 각 존에 따로 존재하고, 서로 다른 존에 분산 배치 가능
- 리전은 그 여러 존들을 묶은 행정 구역 같은 개념


--> 존끼리는 통신이 빠름! 서버끼리도 주고받을 수 있지! 헷갈리지 말자.

<br>

예를 들어, 

서로 다른 리전 A, B에, 각각 메인 사이트와 백업 사이트를 마련함.

이때, 같은 리전 안에서, 여러 개의 존을 선택(multi zone)하여 <span style="color:#3C5087">**sys의 이중화 및 부하의 분산처리**</span>를 가능하게 함.

<span style="color:#3C5087">**sys의 이중화와 부하의 분산**</span>에는 <span style="background-color:#FAFAB4">'로드 밸런서'의 기능을 사용함.</span>

> 로드 밸런서는, 외부의 접속을 통신 데이터양 (예로, 평균 동시 TCP 세션 수가 임계값 초과 시)등의 **조건에 따라 적절한 가상 서버에 전달함.**


--> 통신의 부하를 분산하여 대량 접속에 대응 가능함.

<br>

##### **오토 스케일**

> 접속이 집중되는 때와 같이,
>
>  처리 데이터양에 따른 부하 (서버 CPU나 메모리 사용량이 임계치를 벗어난 경우)가 있으면,
>
> <span style="color:#3C5087">**자동으로 가상 서버의 대수를 증감시키는 기능.**</span>

<br>

##### **정기적인 스냅샷 (백업기능)**

> 수동 또는 이용자가 정한 주기에, 가상 서버 디스크의 <span style="color:#3C5087">**백업을 자동으로 생성함.**</span>

### **클라우드 스토리지 SV**

> 데이터 아카이브(보관), 백업(보호), 파일 서버, 재해대책 등으로 사용되는 대표적인 SV.

클라우드 스토리지는 99.99% 내구성과 99.99% 가용성을 갖춘 SV가 있고,

가용성이 낮지만 크게 비용을 절감할 수 있는 SV가 있는 등, 다양한 종류가 준비됨.



-  다수의 데이터 센터에 데이터를 분산저장하여 데이터 손실을 예방함.

<br>

대표적인 예시: 

- 아마존 웹 서비스(AWS)가 제공하는 `S3(Amazon Simple Storage Service)`
    > 데이터를 **파일 단위**로 저장/검색/삭제하는 저장소.
    > 
    > <span style="background-color:#FAFAB4">'오브젝트 스토리지'</span>라고도 부름.
    >
    > - HTTP/HTTPS를 사용해 데이터를 읽거나 쓸 수 있음.
    > - 웹 서버처럼 정적 콘텐츠를 제공함.

<br>

S3등의 오브젝트 스토리지 SV는 데이터의 읽기 및 쓰기 속도가 느린 편.

--> 따라서 엄격한 응답시간이 요구되는 DB를 쓸 경우, **블록 단위로** 엑세스 할 수 있는 <span style="background-color:#FAFAB4">블록 스토리지</span> SV를 선택함.

대표적 블록 스토리지 SV:

- EBS(Amazon Elastic Block Sotre)

<br>

또, 기업의 파일 서버로 사용하는 경우, 파일 단위로 스토리지 액세스가 가능하고,

파일 공유 기능을 갖춘 SV를 사용함.

대표적인 예시:

- EFS(Amazon Elastic File System)

#### ☁️ 클라우드 네트워크 서비스란?

클라우드 네트워크 서비스는 **클라우드 제공업체가 인터넷을 통해 네트워크 인프라를 제공하고, 사용자가 이를 활용해 자신만의 네트워크 환경을 구성하는 서비스**야.

##### 🔑 핵심 개념

- **클라우드**: AWS, Azure, Google Cloud 같은 사업자가 서버, 스토리지, 네트워크 등 IT 자원을 인터넷을 통해 제공
- **네트워크 서비스**: 사용자가 가상 네트워크, 라우터, 방화벽, VPN 등을 설정해 **자신만의 네트워크 구조를 구성**할 수 있음
- **인터넷 기반**: 모든 자원은 인터넷을 통해 접근하고 관리됨



##### 🧩 예를 들어보면

네가 AWS를 사용한다고 가정해보자:

1. **VPC (Virtual Private Cloud)**: AWS에서 제공하는 가상 네트워크. 네가 직접 IP 범위, 서브넷, 라우팅 테이블 등을 설정해서 **자신만의 네트워크 환경을 구축**할 수 있어.
2. **서브넷**: 퍼블릭 서브넷과 프라이빗 서브넷을 나눠서, 외부와 통신 가능한 영역과 내부 전용 영역을 구분할 수 있어.
3. **인터넷 게이트웨이 / NAT 게이트웨이**: 외부 인터넷과 통신할 수 있도록 연결해주는 장치
4. **보안 그룹 / ACL**: 네트워크 접근 제어를 위한 방화벽 역할



##### 🧠 정리하면

네 말처럼,
> “인터넷을 통해 다른 사업자의 자원을 쓰고, 이 자원을 가지고 나만의 내부 네트워크를 만든다”

는 **클라우드 네트워크 서비스의 핵심을 잘 짚은 표현**이야. 다만, 이 내부 네트워크는 단순히 컴퓨터 몇 대를 연결하는 수준이 아니라, **가상화된 네트워크 인프라 전체를 설계하고 운영하는 것**이야.

### !참고

VPN이 단순한 접속 방식이 아니라 **보안과 인증을 전제로 한 연결 방식**이라는 점을 정확히 짚고 넘어가야 해. 아래에 두 가지 질문을 각각 설명해볼게.

---

#### ❓ 1. VPN을 사용하려면 VPN 게이트웨이를 설치해야 하나?

**✅ 정답: 그렇다.**

- VPN을 통해 네트워크에 접속하려면 **VPN 게이트웨이(또는 서버)**가 반드시 있어야 해.
- 이 게이트웨이는 **VPN 클라이언트의 요청을 받아들이고, 인증하고, 암호화된 터널을 생성**해주는 역할을 해.
- 클라우드에서는 이 역할을 하는 서비스가 예를 들어:
  - AWS: **VPN Gateway**, **Client VPN Endpoint**
  - Azure: **VPN Gateway**
  - GCP: **Cloud VPN**

📌 즉, **VPN 게이트웨이 없이 VPN 접속은 불가능**해. 마치 문이 있어야 열쇠로 들어갈 수 있는 것처럼.

---

#### ❓ 2. VPN은 외부에서 별도 허가 없이 접속하는 방식인가?

**❌ 정답: 절대 아니다.**

- VPN은 **보안이 핵심**이기 때문에, **접속하려면 반드시 인증 절차를 거쳐야 해**.
- 일반적으로 다음과 같은 인증 방식이 사용돼:
  - 사용자 이름 + 비밀번호
  - 인증서 기반 인증
  - MFA (Multi-Factor Authentication, 다중 인증)
- VPN은 **허가받은 사용자만 접속할 수 있도록 설계된 방식**이야.  
  아무나 인터넷만 있다고 접속할 수 있는 게 아니야.

---

#### 🔐 비유로 설명하면

> VPN은 마치 **비밀 통로가 있는 건물** 같아.

- 그 건물에 들어가려면 **비밀문을 열 수 있는 열쇠(VPN 클라이언트 + 인증 정보)**가 필요하고,
- 그 문은 **건물 관리자(VPN 게이트웨이)**가 지키고 있어.
- 아무나 들어올 수 없고, **허가받은 사람만 안전하게 들어올 수 있는 구조**야.

| 질문 | 답변 |
|------|------|
| VPN을 사용하려면 VPN 게이트웨이가 필요한가? | ✅ 반드시 필요함 |
| VPN은 외부에서 아무나 접속할 수 있는가? | ❌ 인증된 사용자만 가능함 |

VPN의 **보안 기능**과 **우회 기능**은 서로 다른 목적에서 사용되지만, 기술적으로는 같은 원리를 활용해. 아래에 그 차이를 명확하게 설명해볼게.

---

#### 🌍 VPN으로 “우회 접속”한다는 의미

VPN을 이용해 **특정 국가에서 차단된 콘텐츠나 사이트에 접속하는 것**을 흔히 “우회”라고 표현해.  
이건 보안 목적이 아니라 **접속 위치를 바꾸는 목적**이야.


#### 🔧 기술적 원리

- VPN을 사용하면 **내 컴퓨터와 VPN 서버 사이에 암호화된 터널**이 생겨.
- 이 터널을 통해 모든 인터넷 트래픽이 **VPN 서버를 거쳐서 나감**.
- 따라서 **웹사이트나 서비스는 내 실제 위치가 아니라 VPN 서버의 위치를 인식**하게 돼.

📌 예시:
> 한국에서 접속하면 차단되는 사이트가 있다고 하자.  
> 미국에 있는 VPN 서버에 연결하면, 그 사이트는 “미국에서 접속했네?”라고 인식해서 접속을 허용함.


#### 🧭 VPN의 두 가지 사용 목적

| 목적 | 설명 | 예시 |
|------|------|------|
| **보안 접속** | 안전하게 내부 네트워크에 접근 | 회사 VPC에 접속 |
| **우회 접속** | 위치를 바꿔서 콘텐츠 제한을 피함 | 넷플릭스 미국 콘텐츠 보기, 특정 국가에서 차단된 뉴스 사이트 접속 |



#### ⚠️ 주의할 점

- **우회 접속은 합법성과 윤리성 문제가 있을 수 있어.**
  - 일부 국가에서는 VPN 사용 자체를 제한하거나 불법으로 간주함
  - 기업이나 서비스 제공자가 VPN 접속을 차단하는 경우도 있음
- **보안 목적의 VPN과 우회 목적의 VPN은 설정 방식과 사용 환경이 다름**
  - 보안 VPN은 인증, 접근 제어가 철저함
  - 우회 VPN은 보통 상용 VPN 서비스(예: NordVPN, ExpressVPN 등)를 사용함

---

#### 🔐 정리하면

> VPN으로 “우회한다”는 건, **내 인터넷 트래픽을 다른 나라의 서버를 통해 보내서, 내 위치를 바꾸는 것**이야.  
> 이건 보안 접속과는 다른 목적이지만, **기술적으로는 같은 VPN 터널링 원리**를 사용해.



---

**IPsec(IP Security)**은 VPN을 비롯한 많은 보안 통신에서 사용하는 **암호화 프로토콜**이야.  
아래에 쉽게, 그리고 깊이 있게 설명해볼게.

---

#### 🔐 IPsec이란?

**IPsec(IP Security)**은 **IP 네트워크에서 데이터를 안전하게 주고받기 위한 암호화 통신 방식**이야.  
즉, 인터넷처럼 **불특정 다수가 사용하는 네트워크에서도 안전하게 통신할 수 있도록 해주는 기술**이지.

---

#### 🧠 핵심 기능

IPsec은 크게 세 가지 기능을 제공해:

| 기능 | 설명 |
|------|------|
| **암호화 (Encryption)** | 데이터를 읽을 수 없도록 암호화 |
| **인증 (Authentication)** | 상대방이 진짜인지 확인 |
| **무결성 (Integrity)** | 데이터가 중간에 변조되지 않았는지 확인 |

---

#### 🧱 IPsec의 구성 요소

##### 1. **AH (Authentication Header)**  
- 데이터의 **무결성과 인증**을 보장  
- 하지만 **암호화는 하지 않음**

##### 2. **ESP (Encapsulating Security Payload)**  
- 데이터의 **암호화 + 무결성 + 인증**을 모두 제공  
- VPN에서는 주로 **ESP를 사용**

---

#### 🔄 동작 방식

IPsec은 두 가지 방식으로 작동할 수 있어:

| 모드 | 설명 | 사용 예시 |
|------|------|-----------|
| **Transport Mode** | IP 패킷의 **데이터 부분만 암호화** | 클라이언트 ↔ 서버 통신 |
| **Tunnel Mode** | IP 패킷 전체를 암호화하고, **새로운 IP 헤더를 붙임** | VPN 터널 (Site-to-Site VPN 등)

📌 VPN에서는 **Tunnel Mode**가 일반적이야. 왜냐하면 **전체 패킷을 보호해야 하니까.**

---

#### 🔐 예시 시나리오

> 회사 본사와 지사가 VPN으로 연결되어 있고, IPsec을 사용한다고 하자.

- 본사에서 보낸 데이터는 IPsec으로 **암호화 + 인증 + 무결성 체크**
- 지사에서는 이를 **복호화하고 검증**
- 중간에 누가 데이터를 가로채도 **내용을 볼 수 없고, 변조도 감지됨**

---

#### 🧠 정리하면

| 항목 | 설명 |
|------|------|
| IPsec이란? | IP 네트워크에서 안전한 통신을 위한 암호화 프로토콜 |
| 주요 기능 | 암호화, 인증, 무결성 |
| 핵심 구성 | AH, ESP |
| 사용 모드 | Transport Mode, Tunnel Mode |
| VPN에서의 역할 | VPN 터널을 안전하게 보호하는 핵심 기술 |

---

### **클라우드 DB SV**

> <span style="color:#3C5087">**OS나 미들웨어는 클라우드 사업자가 관리.**</span>
>
> 사용자는 목적에 부합하는 다양한 DB SV를 선택해 빠르게 사용할 수 있다.

AWS 예시:

#### **1. RDBMS**

> Relational Database Management Sys: 관게형 DB 관리 sys
>
> 관계형 데이터베이스 관리 시스템으로, 데이터를 테이블 형태로 저장하고 `행(row)`과 `열(column)`로 구성됨.

특징
- `스키마 기반`: 데이터 구조가 엄격하게 정의됨 (예: 이름은 문자열, 나이는 숫자)
- `SQL 사용`: 데이터를 조회하거나 수정할 때 SQL(Structured Query Language)을 사용
- `정합성 보장`: 트랜잭션 처리가 강력해서 데이터의 일관성과 무결성을 유지
- `관계 설정 가능`: 테이블 간 외래 키(foreign key)로 관계를 설정


- DB 엔진으로는 MySQL 같은 무료 엔진과 Oracle DB, MS SQl Server 같은 유료엔진 선택 가능함.

- 유료 엔진의 경우, BYOL(Bring Your Own License)이 가능해서, 온프레미스에서 쓰던 제품 라이센스를 그대로 클라우드 서비스로 옮길 수 있음.

-  AWS가 독자적으로 개발한 RDBMS인 `Amazon Aurora`도 있음.


#### **2. NoSQL**

> RDBMS이후 나온 DB SV.
>
> 비관계형 데이터베이스로, 테이블 대신 다양한 형태로 데이터를 저장합니다. 유연한 구조와 확장성이 특징임.

| 유형 | 설명 | 예시 | 
|---|---|---|
| 문서(Document)형 | JSON, BSON 같은 문서 형태로 저장 | MongoDB | 
| 키-값(Key-Value) | 간단한 키와 값 쌍으로 저장 | Redis, DynamoDB | 
| 열(Column)형 | 열 중심으로 저장, 대규모 분석에 적합 | Cassandra, HBase | 
| 그래프(Graph)형 | 노드와 엣지로 관계를 표현, 연결 분석에 적합 | Neo4j | 


특징
- `스키마 없음 또는 유연함`: 구조가 자유로워 빠르게 개발 가능
- `수평 확장에 강함`: 서버를 추가해 쉽게 확장 가능
- `빅데이터, 실시간 처리`에 적합


- `Amazon DynamoDB`는 AWS가 독자 개발한 SV로, 관계형 기능은 없지만, 가용성이 높다.

#### **3. 데이터 웨어하우스**

> 기업의 의사결정을 돕기 위해 다양한 데이터를 통합, 저장, 분석하는 시스템임.


특징

- `분석용 데이터 저장소`: 운영 DB와는 별도로, 분석에 최적화된 구조 = 주제별로 구성됨.
- `ETL 과정 필요`: 데이터를 추출(Extract), 변환(Transform), 적재(Load)해서 저장
- `시간에 따른 변화 추적`: 과거 데이터도 보존하여 트렌드 분석 가능 = 시계열을 가짐.
- `대용량 처리에 최적화`: 수십 테라바이트 이상의 데이터를 효율적으로 관리
- 데이터를 삭제하거나 갱신하지 않음.


<br>

#### **총 비교**

| 항목 | RDBMS | NoSQL | 데이터 웨어하우스 | 
|---|---|---|---|
| 구조 | 테이블 기반 | 문서, 키-값, 그래프 등 | 테이블 기반 (분석 최적화) | 
| 스키마 | 엄격함 | 유연함 | 엄격하지만 분석 중심 | 
| 사용 목적 | 트랜잭션 처리 | 빠른 개발, 유연한 저장 | 분석, 리포팅 | 
| 확장성 | 수직 확장 | 수평 확장 | 수직/수평 모두 가능 | 
| 예시 | MySQL, Oracle | MongoDB, Redis | Redshift, BigQuery |

- AWS의 `VMware Cloud on AWS`

기존 sys을 말 그대로 '**들어 옮긴 후, 최적화하는 방법**'.

4. `애플리케이션 단위의 시프트`
    - sys 전체에 영향을 적게 미치는 애플리케이션을 클라우드로 옮김.

#### 클라우드 ERP

#### ERP란?
> 업의 다양한 부서와 업무를 하나의 통합된 시스템으로 관리하는 소프트웨어를 의미함.
>
>ERP는 기업의 회계, 인사, 재고, 생산, 구매, 판매, 물류, 프로젝트 관리 등 다양한 비즈니스 프로세스를
> 
> 하나의 플랫폼에서 통합적으로 운영할 수 있게 해주는 시스템임.


주요 목적
- 업무 효율성 향상: 중복 작업 제거, 실시간 정보 공유
- 데이터 통합: 부서 간 데이터 일관성 확보
- 의사결정 지원: 정확한 데이터 기반의 분석과 보고
- 비용 절감: 자동화와 최적화로 운영 비용 절감


> 전사적 자원 관리(ERP) 패키지의 기능을 클라우드 환경에서 이용함.
>
> - 자사에 구축하는 사내 ERP에 비해 과도한 커스터마이징을 할 수 없음.
> - --> 개발과 운영의 효율을 높임.
> - --> 비용절감이 가능함.

### **가상화 기술**

> 서버 같은 HW 리소스를 논리적으로 다룰 수 있게 만드는 메커니즘.

- 하나의 **물리 서버 리소스를 여러 개로 나누어**, 여러 개의 서버 환경을 구축할 수 있음.

- 여러 대의 물리 서버 리소스를, **하나의 서버 환경으로 통합**할 수 있음.

--> 결과적으로, <span style="color:#3C5087; background-color:#FAFAB4">**sys 구성을 유연하게 추가/축소 하는 등의 작업이 가능함.**</span>

<br>

#### **서버 가상화**

> 하나의 물리적 서버 리소스에 여러 개의 서버 환경을 할당, 각각의 환경에서 OS와 App를 실행함.
> 

- 이전 버전의 OS에서 동작하는 업무용 App 사용함.

- 다양한 버전의 OS 및 미들웨어 환경이 필요한 개발$\cdot$테스트 환경을 이용함.

각 가상서버는 독립된 상태

--> 서버 중 하나가 위협에 노출돼도 나머지에 영향을 미치지 X

<br>


**3 종류의 서버 가상화 기술**

- `하이퍼바이저 형`

하이퍼 바이저라는 가상화 SW를 동작시킴.

그 위에 Linux, Windows 등 여러 게스트OS를 가동시킴.

<span style="color:#D2691E">당연히 여러 개로 나뉜 서버는 하나의 물리 서버에 비해 성능이 떨어짐.</span>


예시:

a. VMware의 Vsphere

b. MSsoft의 Hyper-V

c. Linux 표준 기능의 KVM 등

- `호스트 OS형`

VMware의 "Workstation"

- `컨테이너 형`

> **서버 자원을 공유한다는 관점에서의 '서버 가상화 기술'**





<서버 가상화 vs 컨테이너>
 
| 항목 | 서버 가상화 (Virtualization) | 컨테이너 (Container) | 
|---|---|---|
| 격리 수준 | 하드웨어 수준 (VM 단위) | 운영체제 수준 (프로세스 단위) | 
| 기술 기반 | 하이퍼바이저 (Hypervisor) | 컨테이너 런타임 (예: Docker, containerd) | 
| 운영체제 | 각 VM마다 별도의 OS 필요 | 호스트 OS 공유 (커널 공유) | 
| 자원 사용량 | 무겁고 자원 많이 사용 | 가볍고 빠름 | 
| 부팅 시간 | 느림 (OS 부팅 필요) | 빠름 (초 단위 실행 가능) | 
| 사용 목적 | 여러 OS 환경 운영, 강한 격리 | 빠른 배포, 마이크로서비스, DevOps | 



하나의 OS 환경에서 app 실행 영역(User Space)를 여러 개로 나누어 사용하는 기술임.

각각의 유저 스페이스를 'container'라고 부름. 각 컨테이너는 별도의 독립된 공간임.

호스트 OS 입장에서는 container마다 하나의 프로세스임.

<span style="background-color:#FAFAB4">! 서버 가상화는 HW 환경을 통째로 가상화함 <--> 컨테이너는 app 실행 환경을 가상화 함.</span>

1. container는 다양한 OS를 지원하는 템플릿을 사용할 수 있음. 

> --> <span style="color:#3C5087">**하나의 host OS에서 멀티 OS 환경을 제공할 수 있음.**</span>

2. 가상 서버는 기동에 수십 초~ 수 분 걸림

> <--> container는 가상화에 대한 오버헤드가 적으므로, <span style="color:#3C5087">**빠르게 기동/정지 가능함.**</span>

3. 게스트 OS가 필요 없기 때문에, <span style="color:#3C5087">**보조 기억장치의 사용량도 줄임.**</span>

4. 개별 컨테이너가 사용하는 HW 리소스가 적음

> --> <span style="color:#3C5087">**1대의 물리 서버에 많은 양의 container를 올릴 수 있음.**</span>

app 실행 환경이 container 단위로 패키징됨.

- 로컬 컴퓨터의 개발환경을 클라우드로 옮길 수도 있음.
- 다른 클라우드로 이식이 쉬움.

예시:

대표적인 컨테이너 타입 가상화 SW: `Docker`

운영 관리 자동화 도구: `Kubernetes`

"`도커(Docker)`는 소프트웨어를 컨테이너라는 표준화된 유닛으로 패키징하여 배포하고 실행할 수 있게 해주는 플랫폼입니다. <br>도커를 사용하면 애플리케이션과 그 의존성을 함께 묶어 어디서든 일관되게 실행할 수 있습니다."

### **분산처리 기술**

> 대량의 데이터를 **여러 서버에 분산해 동시에 병렬로 처리**함.

클라우드 등장 이전, 테라$\cdot$페타B 급 대용량 데이터 처리에는 고속 CPU와 대용량 메모리가 탑재된 서버가 필요했음.

--> <span style="color:#3C5087">**지금은 분산 처리 기술 + 클라우드 SV를 이용해서, 데이터를 여러 개의 서버에 나눠 병렬로 처리가능함.**</span>

- 여러 개의 서버를 결합해 하나의 컴퓨터로 보이게 만드는 기술 = **클러스터링**이라고 부름.

> 대용량 데이터 처리 중 몇 개의 서버에 장애가 발생해도, 나머지 서버에 자동으로 작업을 할당하고 이어감. 

**| 분산 처리를 구현하는 SW**

- Apache Hadoop
    > - 1대의 **master 서버**와 **여러 대의 slave 서버**로 구성.
    >
    >  - 마스터 서버는 데이터 처리 전체제어, 슬레이브는 연산을 처리.
    >
    > - 처리 능력은 슬레이브 서버의 대수에 비례해 늘어남.

- Apache Spark
  > - **메모리 안에서** 대용량 데이터의 병렬 분산 처리를 실행함.
  >
  > - **반복 처리** 중, 디스크에서 데이터를 읽지 않아 빠름.
  >
  > - <span style="color:#D2691E">단, 테라 바이트 급 이상의 처리는 불가함.</span>

### **DB 기술**

> 데이터 분석 처리 및 *트랜잭션 처리*(대량 데이터의 집계, 상품 거래, 빅 데이터 분석, IoT 기반 등)의 사용 목적별,
>
> RDB와 NoSQL로 대표되는 데이터 특성별로 제공됨.

**! 참고**

트랜잭션 처리란,

>"데이터베이스나 시스템에서 일련의 작업들을 하나의 단위로 묶어서 처리하는 방식을 말해요. 
>
>이 단위는 모든 작업이 성공해야만 전체가 성공으로 간주되고, 하나라도 실패하면 전체를 되돌리는 방식이에요. 
>
>쉽게 말해, "모두 아니면 아무것도"라는 원칙이죠."


트랜잭션의 핵심 개념: ACID

트랜잭션이 제대로 작동하려면 다음 네 가지 속성을 만족해야 해요:

| 속성 | 설명 | 
|---|---|
| Atomicity (원자성) | 트랜잭션은 모두 실행되거나 전혀 실행되지 않아야 함 | 
| Consistency (일관성) | 트랜잭션 전후에 데이터는 항상 일관된 상태여야 함 | 
| Isolation (격리성) | 동시에 실행되는 트랜잭션은 서로 영향을 주지 않아야 함 | 
| Durability (지속성) | 트랜잭션이 완료되면 그 결과는 영구적으로 저장되어야 함 | 






주요 DB는 RDB나 NoSQL 등이 있음.

**| RDB**

> 여러 개의 데이터를 행과 열이 있는 <span style="color:#3C5087">**표 형식으로 표현**</span>, 복잡한 데이터의 관계를 잘 처리할 수 있도록 만든 DB.

RDBMS라는 전용 관리 SW를 이용한다.

예를 들면,

|키|이름|소속|
|---|---|---|
|1|윤경|영업부|
|2|연희|업무부|

데이터 조작언어: SQL

데이터를 일관적이고 엄격하게 관리.

DB처리 능력이 HW기능 강화(**스케일업**)에 달림.

`Oracle DB`, `MySQL`, `MS SQL Server`, `PostrgeaSQL` 등등

<br>

<br>

**| NoSQL**

> RDB가 아닌 DB.

제품별로 <span style="color:#3C5087">**다양한 자료구조**</span>를 제공함.

<span style="color:#3C5087">**대량의 데이터를 분산시켜 고속으로 처리하는 분산 DB라는 공통특징**</span>이 있음.

분산시켜 처리한다는 특징이 있어 클라우드 SV에 적합함.

NoSQL의 자료구조의 차이에 따라 

- 키 밸류 형
- 컬럼 지향형
- 문서 지향형
- 그래프형으로 분류함.

데이터 조작 언어는 제품마다 다름.

일시적으로 데이터의 일관성이 없는 상태가 있음(결과 무결성).

DB처리 능력이 **스케일 아웃**(서버 대수 증가)와 큰 관련있음.

**키 밸류 형**
> 제품의 모든 데이터는 인덱싱된 값으로 구성됨.
>
> 구성이 간단, 확장성 높음, 데이터 읽기 속도 빠름.

`Memcached`, `Redis`, `Riak` 등

|키|벨류|
|---|---|
|k1|aaa,bbb,ccc|
|k2|aaa,bbb|




**컬럼 지향형**
> 컬럼 단위로 데이터를 보관함.
>
> 대량의 컬럼 단위 데이터의 집계/갱신 속도가 빠름.
>
> 데이터의 쓰기 속도가 빠름.

`Hbase`, `Cassandra` 등

|컬|럼|지|향|
|---|---|---|---|
|key |\|   |column|   |  
|   |\| id|\| name|\| tiemstamp|
|k001|u001|stephan|123456|
|k002|u002|Jackson|234567|
|k003|u003|Terry|345678|



**지향형**
> 복잡한 데이터를 문서에 저장하고 문서 단위로 데이터를 저장하여, 검색하고 관리함.
>

```JSON
{
    "ID" : ObjID( "AAA" )
    "title": "클라우드의 기본"
    ...
}

**그래프형**
> 데이터 간의 관계를 그래프로 형성함.
>
> 통합 검색을 빠르게 할 수 있음.

`Neo4j` 등

노드, 관계, 속성이라는 3가지 요소로

노드 간의 관계를 표현하는 자료 구조.

<br>

### **스토리지 기술**

> 스토리지는 데이터와 프로그램을 저장하는 기록 장치임.

<br>

**| 블록 스토리지**

> 일정한 크기의 블록으로 나뉜 스토리지의 논리 볼륨을, **블록 단위로 액세스**할 수 있는 방식임.

스토리지 액세스에는 <span style="background-color:#FAFAB4">파이버 채널(FC)와 iSCSI 같은 전용 프로토콜을 사용함.</span>

<span style="color:#3C5087">**서버와 스토리지가 데이터를 교환할 때의 오버헤드가 적음.**</span>

--> <span style="color:#3C5087">**빠른 데이터 전송이 가능함.**</span>

--> 낮은 ratency가 요구되는 DB용도로 사용되고 있음.

<span style="color:#D2691E">기존 저장소 다 써서 추가하려면 머리 아픔... RAID나 SAN 신경써야 함.</span>



**! 참고**

`RAID` (Redundant Array of Independent Disks): 여러 개의 하드디스크를 하나의 논리적 볼륨으로 묶는 기술이에요.
- 목적: 속도 향상, 데이터 보호, 고가용성
- 예시:
- RAID 0: 빠르지만 복구 불가
- RAID 1: 복제(미러링)으로 안정성 확보
- RAID 5/6: 패리티 기반으로 일부 디스크 장애에도 복구 가능
- RAID 10: 성능과 안정성 모두 확보
- NAS에서도 RAID 설정이 중요해요. 자세한 RAID 설명은 이 블로그에서 확인할 수 있어요.

`SAN` (Storage Area Network): 고속 네트워크를 통해 스토리지를 서버에 연결하는 방식이에요.
- 서버는 SAN을 통해 마치 로컬 디스크처럼 스토리지를 사용함.
- 주로 기업용 데이터센터에서 사용되며, 블록 스토리지 기반이에요.
- 설정은 복잡하고 고가의 장비가 필요하지만, 성능은 매우 뛰어나요.


<br>

**| 파일 스토리지**

> 파일을 그대로 읽고 쓰는 공유할 수 있는 스토리지임.

데이터 처리는 

- Windows OS에서 사용하는 `SMB(Server MessageBlock)`
-  `CIFS`(Common Internet FileSystem)
-  UNIX및 Linux OS에서 사용하는 `NFS`(Network FileSystem)


등의 **파일 공유 프로토콜을 사용**하여 **파일 단위로 이루어짐.**

주로 파일 서버로 이용하며, <span style="color:#3C5087">**파일의 접근 제어나 파일의 속성 정보를 관리하기 쉽다**</span>는 장점이 있음.

- 디렉토리 기반 계층 구조로 관리됨.
- 수많은 파일이 쌓이면 디렉토리 탐색 속도가 느려지고 성능 저하.
- NAS 장비나 파일 서버의 용량에 따라 확장 한계가 존재.


**! 참고**

`NAS (Network Attached Storage)`는 네트워크에 연결된 저장 서버.
- 여러 사용자가 동시에 접근 가능.
- 파일 공유, 백업, 미디어 서버, CCTV 저장 등에 사용됨.
- 자체 운영체제(OS)를 탑재하고 웹 GUI로 관리 가능.
- RAID 설정을 통해 데이터 보호 가능.
- 대표 브랜드: Synology, QNAP, TrueNAS 등
- 자세한 설명은 이 블로그에서 볼 수 있어요.




<br>

**| 오브젝트 스토리지**

> 데이터를 객체 단위로 처리함.

데이터 및 관련 메타 데이터로 구성된 **오브젝트**에는 **고유한 ID(URI)가 부여**됨.

오브젝트 스토리지는 <span style="color:#3C5087">**OS나 F.sys에 의존하지 않음.**</span>

그러면서 데이터를 저장하고 오브젝트에 엑세스할 수 있음.

--> 뭔소리냐면, <span style="background-color:#FAFAB4">각 오브젝트를 분산해서 저장한다</span>는 소리임. 

**데이터+메타데이터+ID단위**로 묶여서, 각 묶음은 분산돼 저장함.

오브젝트 스토리지의 액세스에는 **HTTP 프로토콜** 기반의 **REST(REpresentational State Transfer)형식 API를 사용**함.

- <span style="color:#3C5087">**쉽게 용량을 늘릴 수 있음.**</span>

- <span style="color:#3C5087">**데이터의 크기와 저장할 수 있는 데이터의 수에 제한 X.**</span>

--> 갱신 빈도가 적은 데이터나/ 대량의 데이터를 저장, 장기 보존하는 용도로 이용되고 있음.

**! 참고**

스토리지 방식 비교
| 항목 | 오브젝트 스토리지 | 블록 스토리지 | 파일 스토리지 | 
|---|---|---|---|
| 구조 | 데이터 + 메타데이터 + 고유 ID | 고정 크기 블록으로 분할 | 계층적 디렉토리 구조 | 
|  저장 단위 | 오브젝트 | 블록 | 파일 | 
|  메타데이터 | 풍부하게 커스터마이징 가능 | 없음 또는 제한적 | 제한적 | 
|  접근 방식 | HTTP/HTTPS API 기반 | 로우 레벨 디스크 접근 | 파일 시스템 접근 (NFS, SMB 등) | 
|  확장성 | 매우 뛰어남 (수십 페타바이트 이상) | 제한적 (디스크 크기 기반) | 중간 수준 | 
|  데이터 무결성 | 자동 관리 (버전 관리, 체크섬 등) | 직접 관리 필요 | 일부 지원 | 
|  비용 | 저렴 (대용량 저장에 유리) | 고가 (고성능에 적합) | 중간 수준 | 
|  지연 시간 | 높음 (비동기 처리) | 낮음 (빠른 I/O) | 중간 | 
|  사용 사례 | 백업, 이미지/영상 저장, 빅데이터 | 데이터베이스, VM 디스크 | 일반 문서 저장, 공유 폴더 | 






"기존 저장소가 부족해서 추가한다"는 개념은 같지 않아?

개념적으로는 비슷하지만, 방식이 다릅니다.

| 항목 | 오브젝트 스토리지 | 블록/파일 스토리지 | 
|---|---|---|
| 저장 방식 | 새로운 오브젝트를 분산된 저장소에 자동으로 추가 | 기존 디스크가 꽉 차면 물리적 디스크를 추가해야 함 | 
| 확장 방식 | 시스템이 자동으로 확장 (클라우드 기반) | 수동으로 디스크 추가, RAID/SAN/NAS 설정 필요 | 
| 관리 복잡도 | 낮음 (자동화됨) | 높음 (물리적 구성, 설정 필요) | 


즉, 오브젝트 스토리지는 "추가"가 아니라 "자동 확장"에 가까워요. 반면 블록/파일 스토리지는 사람이 직접 디스크를 추가하고 설정해야 하죠.


<br>

### **번외 - 오픈소스 클라우드 기술**

#### **IaaS를 위한 오픈소스 기술**




> 클라우드 기반 Sw는 <span style="background-color:#FAFAB4">다양한 가상화 및 클라우드 기능을 제공하는 SW.</span>
>
> KVM, VMware, ESXi 등의 **hypervisor 에서 실행함.**

진정한 자체 클라우드를 구축하려함

 --> `OpenStack`으로 대표되는 오픈 소스 클라우드 기반 SW가 대안.

MSsoft의 `Azure`, Google의 `Google Cloud Platform` 등을 제외한 대부분의 <span style="background-color:#FAFAB4">주요 퍼블릭 클라우드 서비스의 IaaS 기반</span>임.

<br>

**| OpenStack**

NASA의 독자적 클라우드 기반 Nebula를 바탕으로 나옴.

**철학**: 특정 업체에 점거되지 X, 업계 표준 사양의 클라우드 인프라 기반을 개발함.

- 표준 개발 언어 **python**
- 표준 외부 API는 OpenStack의 독자적 API: **REST API**및 Amazon EC2/S3 호환
- 표준 OS는 **Ubuntu Linux**

단독으로 동작하는 수많은 SW로 구성됨.

- hypervisor제어 및 베어메탈 프로비저닝을 수행하는 `Noca`
- 가상 NW를 제어하는 `Neutron`
- 운영관리도구 `Horizon` 등

---
! 참고

**프로비저닝 (Provisioning)** 은 IT 시스템에서 필요한 자원(서버, 스토리지, 네트워크 등)을 미리 준비하고 설정하여 즉시 사용할 수 있도록 만드는 과정임.

---

**OpenStack 구조도**

한 눈에 볼 수 있게, OpenStack을 구조화해서 설명해보면,

hypervisor 오버제어(전반적 제어)를 위한 '기반 SW'인 (+ 서버 가상화 SW) `Nova`가 있고,

여기에 플러그인을 통해 다양한 기능을 추가한다.






예를 들면, 

- 스토리지 가상화 SW인 `Swift/Cinder`를 추가해 물리 스토리지 위에 가상 스토리지를 추가하거나,

- NW 가상화 SW인 `Neutron`을 추가해 물리 NW 위에 가상 NW를 추가한다.

<br>

#### **PaaS를 위한 오픈소스 기술**

오픈소스 IaaS기반 SW와 독립적으로 동작, (꼭 이걸로 안해도 됨.)

AWS나 VMware, OpenStack 같은 어느 클라우드 SV에서든 실행시킬 수 있음.

*예시*

`클라우드 파운드리` 등

오픈소스 웹 app 실행 환경을 지원함.

- Ruby, Java, Python 등 다양한 **개발 언어**
- Ruby on Rails, Sinatra, Node.js, Spring Framework 와 같은 오픈소스 표준에 따르는 **개발 프레임워크**
- MySQL, MongoDB 등의 여러 **DB**를 사용할 수 있음.

\+ 컨테이너 지원도 이루어지고 있음.

<-- 내부적으론, 개발자(이용자)가 코드를 보내면, PaaS 기반 SW에서는

1. app 코드 해석
2. 실행 환경이 포함된 컨테이너로 보냄
3. 해당 컨테이너에서 동작시킴.

또, 일반 user가 URL이나 주소로 접속하면, 

*라우터*나 *로드밸런서*가 컨테이너와 URL을 매핑해 **부하를 분산**시키는 방법 사용함.

<br>

**| NW 가상화**

- `VLAN`

> (Virtual LAN). <span style="background-color:#FAFAB4">하나의 물리적인 NW를 여러 개의 논리적 NW로 분할하는 기술.</span>

물리적인 배선을 변경X 

--> **NW 장비(스위치 등)에 설정을 추가해 NW를 나눔.**

논리적으로 나눈 NW들은 라우터를 거치지 않으면 통신 X

--> 조직 단위로 NW를 나눠, 조직 한정 내부 NW를 만들 수 있음.

<br>

- `VPN`

> (Virtual Private Network). <span style="background-color:#FAFAB4">가상으로 전용선과 같은 사설망을 연결하는 기술 (일종의 전용 터널 뚫기).</span>


온프레미스 sys이 인터넷을 통해 VPN에 연결할 때는 IPsec, wireGuard 등의 다양한 암호화/ 인증 등을 보장하는 통신 구현.

<br>

- `NFV`

> (Network Functions Virtualization). <span style="background-color:#FAFAB4">NW 기능을 SW로 구현하여 가상 서버위에 구현하는 기술.</span>

기존 NW장비(방화벽, 라우터, 스위치 등)들은 SW와 HW가 일체화돼있음. 

--> **NW기능을 가상화 기반에서 SW로 구현한다.**

( 기존 NW기능 대부분이 NW 전용 HW와 통합된 NW 어플라이언스로서 제공됐음.

--> 라우터, 게이트웨이, 방화벽, 로드 밸런서와 같은 NW장비의 기능을 가상 서버에서 app SW로 구현함. )

<span style="color:#3C5087">**NW 장비의 수요 및 구성 변경 등에 유연하게 대응가능함.**</span>

<br>

### **SDN**

서버 가상화, 클라우드 발달에 따라 sys의 통합 관리 및 운영 자동화

--> <span style="color:#D2691E">NW 트래픽의 급속한 증감 및 경로 변경을 초래.</span>




**But**, 

기존과 같은 방식으로 **HW별로 운영/관리되고 있음.**

**NW 증설과 변경, 운영의 자동화가 큰 과제임.**

--> 

<span style="background-color:#FAFAB4">**'SDN(Software Defined Networking)'**</span>
- NW가상화
- NW 구성과 기능 설정 등



<br>

`SDN`은,

기존 NW 장비의 '전송 기능(데이터 플레인)'과 '제어 기능(컨트롤 플레인)'을 **분리**함. 

**제어 기능**을 **컨트롤러에 논리적으로 집중**시켜 **데이터의 흐름을 SW로 정의**하는 개념.

기존엔 기능이 2개 다 있어서, 관리자가 1대마다 기기를 설정 

--> 통신 경로를 제어해야했음. &

통신의 종류마다 경로도 분리해야함.

<span style="background-color:#FAFAB4">이제는 SDN 컨트롤러가 컨트롤을 다 맡아 하기 때문에, 얘만 관리 app으로 조작해주면 됨.</span>


---
! 참고

데이터 플레인(Data Plane)은 컴퓨터 네트워크나 시스템 아키텍처에서 자주 등장하는 개념인데, 간단히 말하면 **실제 데이터가 흐르고 처리되는 경로와 기능을 담당하는 부분**이야. 이를 더 잘 이해하려면 **컨트롤 플레인(Control Plane)**과 비교해서 설명하는 게 좋아.

##### 데이터 플레인 vs 컨트롤 플레인

| 구분 | 데이터 플레인 (Data Plane) | 컨트롤 플레인 (Control Plane) |
|------|-----------------------------|-------------------------------|
| 역할 | 실제 데이터 전송 및 처리 | 네트워크 경로 설정, 정책 결정 |
| 예시 | 패킷 전달, 포워딩, 암호화 | 라우팅 테이블 생성, QoS 설정 |
| 위치 | 라우터/스위치의 하드웨어 또는 커널 | 운영체제, 네트워크 OS, 관리 소프트웨어 |
| 속도 | 빠르고 실시간 | 상대적으로 느림, 비주기적 |


##### 예시로 이해해보자

- 네트워크 라우터를 생각해보자.
  - **데이터 플레인**은 네트워크 패킷을 받아서 목적지로 빠르게 전달하는 역할을 해.
  - **컨트롤 플레인**은 어떤 경로로 패킷을 보내야 할지 결정하고, 라우팅 테이블을 업데이트하는 역할을 하지.


#####  어디서 쓰일까?

- **클라우드 컴퓨팅**: Kubernetes에서 데이터 플레인은 실제 애플리케이션 컨테이너가 돌아가는 부분이고, 컨트롤 플레인은 클러스터 상태를 관리하는 부분이야.
- **SDN(Software Defined Networking)**: 컨트롤 플레인은 중앙에서 네트워크 정책을 설정하고, 데이터 플레인은 실제 트래픽을 처리해.
- **보안 시스템**: 방화벽이나 IDS에서 데이터 플레인은 패킷을 검사하고 차단하는 역할을 해.

---

서버 가상화처럼, 물리적 NW가 추상화됨.

--> 하나의 물리적 NW위에 컨트롤러 별로 <span style="color:#3C5087">**여러 개의 가상 NW 구축가능함.**</span>

SDN 등장 이후로,

스위치, 라우터 같은 NW 장비의 아키텍처가,


(OS, 미들웨어, app이 통합된) **수직 통합형 아키텍처**

--> 각 기능별 레이어를 분리해서 **오픈형 I/F로 연결하는 아키텍처**로 바뀌고 있음.

NW 장비를 컨트롤러가 집중적으로 제어하게 되며, 

NW 가동 상황 및 운영에 맞추어 SW로 유연하게 데이터 전송 경로를 변경할 수 있음.

--> 가상서버에서 실행중인 OS나 SW를 정지시키지 않고, 다른 서버로 옮기는 <span style="color:#3C5087">**'라이브 migration'이 쉬워짐.**</span>

---

#### ! 참고

좋은 질문이야! 😊  
**라이브 마이그레이션(Live Migration)**은 가상 머신(VM)을 중단 없이 한 물리 서버에서 다른 서버로 옮기는 기술인데, 이게 **SDN(Software Defined Networking)** 덕분에 훨씬 더 **유연하고 안정적으로** 가능해졌어. 왜 그런지 하나씩 풀어볼게.




##### 라이브 마이그레이션의 핵심 과제

라이브 마이그레이션을 하려면 다음과 같은 네트워크 문제가 생겨:

- VM이 이동하면 **IP 주소나 MAC 주소가 바뀔 수 있음**
- 기존 네트워크 경로가 **VM의 새 위치를 모름**
- VM이 사용하는 **방화벽, QoS, ACL 정책**도 같이 옮겨야 함
- 마이그레이션 중에 **패킷 손실이나 지연**이 생기면 안 됨

이런 문제를 해결하려면 네트워크가 **동적으로 재구성**되어야 해. 그런데 기존의 하드웨어 중심 네트워크에서는 이게 어렵고 느려.



##### SDN이 해결해주는 방식

SDN은 네트워크를 **소프트웨어적으로 제어**할 수 있게 해줘서, 라이브 마이그레이션에 딱 맞는 환경을 제공해:

| SDN 기능 | 라이브 마이그레이션에 어떻게 도움 되는지 |
|----------|------------------------------------------|
| **중앙 집중형 컨트롤 플레인** | VM의 위치 변경을 실시간으로 감지하고, 네트워크 경로를 즉시 재설정 |
| **동적 정책 적용** | VM이 이동해도 방화벽, QoS, ACL 같은 정책을 자동으로 따라옴 |
| **가상 네트워크 추상화** | VM이 어디에 있든 동일한 네트워크 환경을 유지 가능 |
| **트래픽 엔지니어링** | 마이그레이션 중에도 최적의 경로로 패킷 전달, 지연 최소화 |


##### 예시: OpenFlow 기반 SDN

- VM이 서버 A에서 서버 B로 이동하면,
- SDN 컨트롤러가 이를 감지하고,
- OpenFlow 스위치를 통해 **패킷 포워딩 경로를 즉시 변경**해줘.
- 사용자 입장에서는 VM이 이동했는지조차 모를 정도로 **끊김 없이 서비스가 유지**돼.



##### 요약

> SDN 덕분에 네트워크가 VM의 이동을 **실시간으로 인지하고**, 필요한 설정을 **자동으로 조정**할 수 있어서 라이브 마이그레이션이 **끊김 없이, 안전하게** 가능해지는 거야.

---

\+

**QoS**와 **ACL**은 네트워크에서 **트래픽을 제어하고 보호**하는 데 쓰이는 중요한 정책들이야. VM이 이동할 때 이 정책들도 따라가야 하는 이유를 이해하려면 먼저 각각이 뭔지부터 알아야겠지?


##### QoS (Quality of Service)

**QoS는 네트워크 품질을 보장하는 기술**이야. 모든 트래픽이 똑같이 중요한 건 아니니까, 중요한 트래픽을 우선 처리하도록 설정하는 거야.

##### 예를 들어:
- **영상 회의**나 **VoIP 통화**는 지연에 민감하니까 **우선순위 높게**
- **백업 작업**이나 **파일 다운로드**는 지연돼도 되니까 **우선순위 낮게**

##### QoS가 하는 일:
- **대역폭 할당**: 특정 서비스에 최소 대역폭 보장
- **우선순위 지정**: 중요 트래픽 먼저 처리
- **지연/손실 제어**: 패킷 손실이나 지연을 최소화

##### VM이 이동하면?
QoS 정책도 새 위치에서 **동일하게 적용**돼야 영상 통화나 서비스 품질이 **끊기지 않고 유지**돼.



##### ACL (Access Control List)

**ACL은 네트워크 접근을 제어하는 규칙 목록**이야. 누가 어떤 트래픽을 보낼 수 있는지, 받을 수 있는지를 결정해.

#####  예를 들어:
- "이 VM은 80번 포트(HTTP)만 허용하고, 나머지는 차단"
- "내부망에서만 접근 가능, 외부 IP는 모두 차단"

#####  ACL이 하는 일:
- **IP 주소 기반 허용/차단**
- **포트 번호 기반 제어**
- **프로토콜 기반 필터링 (TCP, UDP 등)**

#####  VM이 이동하면?
ACL 정책도 같이 옮겨야 **보안이 유지**돼. 안 그러면 VM이 갑자기 외부에서 접근 가능해질 수도 있어 😱



##### 왜 SDN이 중요하냐면…

SDN이 없으면 VM이 이동할 때 QoS나 ACL을 **수동으로 다시 설정**해야 해.  
하지만 SDN이 있으면 **컨트롤러가 자동으로 정책을 재적용**해줘서,  
VM이 어디에 있든 **동일한 품질과 보안**을 유지할 수 있어.

---

<br>

### **엣지 컴퓨팅**

> 서버를 클라우드 쪽에 배치하는 대신, 서버를 스마트폰 등의 <span style="background-color:#FAFAB4">말단 디바이스에 가까운 영역(엣지)에 분산 배치함.</span>
>
> 그 서버에서 말단 디바이스가 보낸 데이터를 처리함.

말단 디바이스에서 보낸 데이터는 근처 기지국을 통해, 가까운 거리에 있는 서버들에서 처리한다.

클라우드와 상호 보완관계에 있는 기술로서, 말단에서 처리하다보니,

- 낮은 지연
- 실시간 데이터 처리
- 높은 신뢰성

예시

- 공장라인 기계 제어
- 커넥티드 카, 자율 주행차의 영상 데이터와, 주행, 지도 데이터 처리

데이터 수집용 센서 + CPU와 GPU를 탑재해 처리를 실현한 **엣지 컴퓨팅용 디바이스**등장.

<br>

### **하이퍼 컨버지드 인프라스트럭처**

> 클라우드 SV의 기본 기능을 패키지로 제공하는 제품.

`컨버지드 인프라(Converged Infra.)`란, 서버, NW, 스토리지, SW(hypervisor 및 운영관리도구)를 하나의 패키지에 통합한 제품임.

`하이퍼 컨버지드 인프라`란, **SW기반의** 서버와 NW, 스토리지 등의 구성 요소가 통합된 제품임.

컨버지드 인프라와 비슷한 개념을 '가상화'로 제공한다는 차이점이 있음.

<br>

### **GPU**

| CPU | GPU |
|:---:|:---:|
|코어(연산장치)를 수 개~수십 개|코어(연산장치)를 수십 개~수천 개|
|분기에측/명령 스케쥴링 등 기능을 가진 회로가 많음|같은 작업을 여러 개 코어에 할당하는 병렬 처리를 할 수 있음|
|OS와 같은 복잡한 프로그램 처리와 범용적인 프로그램 처리| 대량의 계산이 요구되는 딥러닝, 물리 시뮬레이션 등의 처리|

<br>

### **데이터센터**

- 안전한 건물
- 출입 관리 엄격
- 댈야의 서버와 NW 장비 등의 운영과 공조sys 관리를 위한 전력공급

전력 효율성을 나타내는 지표 PUE(Power Usage Effectiveness)

$PUE = 데이터\ 센터\ 전체의\ 소비\ 전력\ \div\ IT\ 기기별\ 소비전력$

1에 가까울수록 전력 효율이 높다. 현재 표준 데이터센터는 1.8~2.0 (19?20?년 기준)

<br>

### **serverless architecture**

| 개요

`마이크로 서비스 아키텍처`라는게 보급되고 있음.

- 하나의 app을 작은 SV의 집합체로 구축함.

- 개별 SV들이 API와 같은 간단한 방법으로 연계해서 동작함.

서버와 스토리지, DB, NW 등의 기능들이 `컴포넌트`라는 형태의 독립된 SV로 제공되며,

컴포넌트끼리 **느슨하게 결합된 상태(API)로 전체 기능을 구현**함.

--> 개인이 각 컴포넌트를 조합하고 API로 연계해 독자적 app개발이나 SV, sys 구축이 가능해짐.

--> 이때 클라우드 SV가 풀 매니지먼트(환경 구축 및 보안 패치, 백업 검색 등을 모두 클라우드 사업자가 수행)하는 경우,

사용자는 서버의 존재를 전혀 의식하지 않고 app을 동작시킴.

이를 **서버리스 아키텍처**라고 부름.

| 정리


`온프레미스`와 `클라우드 가상서버`(서버의 HW는 사업자가, 서버 프로그램은 사용자가 관리) = 서버의 존재를 의식함

`클라우드 풀 매니지먼트 SV`(서버 HW부터 프로그램 모두 사업자가 관리함. 따라서 기능만 씀) = 서버의 존재를 신경쓰지 않음(serverless - )

___

#### ! 참고

**API**

대다수 클라우드 SV에는 API가 마련됨.

> API는 프로그램이 가진 **기능이나 리소스**를, **외부의 다른 프로그램이 호출**하여 이용하기 위한
>
> **명령이나 함수, 데이터 형식** 등을 **정한 규약**임.
>
> 

서버의 예를 들면, 

가상 서버에 마련된 API를 사용하여 프로그램을 작성하면, 

가상 서버의 생성과 정지 같은 조작을 사람의 손을 거치지 X

--> 프로그램으로 직접 제어할 수 있게 됨.

클라우드 SV에서는 복잡한 API를 조작하기 위한 명령줄 도구와 개발용 SDK(Software Development Kit)가 마련돼 있음.




<br>

<br>

<br>

<br>

---

# **| 기타 |**

<br>

## **클라우드 SV에 대응하는 각종 app**

app은 클라우드에 대한 대응 정도에 따라 3가지로 나뉜다.

1. `트레디셔널 app`
    - 기업에서 온프레미스로 운용하는 기존 sys과 각종 서버들
    - ITIL에 의한 (IT Infra. Library) 운용 관리(사내 레거시로 내려오는 지침들)
2. `클라우드 대응 트레디셔널 app`
    - 프라이빗 클라우드 및 클라우드 사업자의 클라우드 SV 위에서 운용되는 기존 sys.
    - 트레디셔널 app과 같은 설꼐 사상
    - ITIL에 의한 운용관리
3. `클라우드 네이티브 app`
    - 클라우드 SV에서의 이용을 전제함.
    - 마이크로 SV로 설계됨.
    - API를 통해 외부 SV와 원활히 연계,  API를 활용한 애드온(확장 기능)개발도 쉽도록 설계된 app
    - DevOps를 통한 운영과 관리

<br>

## **하이브리드 클라우드의 구성**

> **하이브리드 클라우드**는 
>
> **온프레미스 인프라**(자체 서버)와 **퍼블릭 클라우드 서비스**(AWS, Azure, Google Cloud 등)를 **결합한 IT 환경**임. 
>
> 이 구성은 유연성과 확장성을 높이면서도 보안과 비용 효율성을 고려할 수 있도록 설계됨.


#### **| NW 연결을 통한 연계**

하이브리드 클라우드를 구성할 때는,

**클라우드 SV**와 **데이터 센터**, 그리고 **온프레미스 sys**를 <span style="background-color:#FAFAB4">VPN망, 또는 전용선으로 연결</span>

이때, AWS DIrect Connect나 Azure ExpressRoute같은 전용선으로 연결하는 접속 SV를 많이 이용함.


--> 온프레미스와 퍼블릭 클라우드 SV사이를 <span style="color:#3C5087">**보안성 확보 + 낮은 레이턴시 환경 구축 가능함.**</sspan>

#### **| 클라우드 관리 플랫폼을 통한 통합**

여러 개의 클라우드 SV를 포털 화면으로 API를 통해 관리하고 제어함.

대부분의 클라우드 관리 플랫폼은 AWS, Azure, VMware, OpenStack과 연계 가능해서 

--> 다양한 퍼블릭 클라우드와 프라이빗 클라우드를 **통합**해 **통합 클라우드 sys를 운용**할 수 있음.

다음은 하이브리드 클라우드의 주요 구성 요소입니다:



#### ☁️ 하이브리드 클라우드 구성 요소

##### 1. **온프레미스 인프라 (Private Cloud 또는 자체 데이터센터)**
- 기업 내부에 위치한 서버, 스토리지, 네트워크 장비
- 민감한 데이터나 핵심 업무를 처리
- VMware, OpenStack 등을 활용해 프라이빗 클라우드로 구성 가능

##### 2. **퍼블릭 클라우드 서비스**
- 외부 클라우드 제공업체가 운영 (예: AWS, Microsoft Azure, Google Cloud)
- 탄력적인 리소스 확장 가능
- 비핵심 업무, 테스트 환경, 백업 등에 활용

##### 3. **클라우드 관리 플랫폼 (CMP) 또는 오케스트레이션 도구**
- 두 환경을 통합 관리하는 도구
- 예: Microsoft Azure Arc, Google Anthos, Red Hat OpenShift
- 리소스 모니터링, 보안 정책 적용, 자동화 기능 제공

##### 4. **네트워크 연결 및 보안**
- VPN, 전용 회선, SD-WAN 등을 통해 온프레미스와 퍼블릭 클라우드를 연결
- 보안 게이트웨이, IAM(Identity and Access Management), 암호화 기술 적용

##### 5. **데이터 및 애플리케이션 통합**
- 데이터 동기화 및 마이그레이션 도구
- API 게이트웨이, 컨테이너 기술(Docker, Kubernetes)로 애플리케이션을 유연하게 배포



#### 🔄 하이브리드 클라우드의 활용 예시

| 사용 사례 | 온프레미스 | 퍼블릭 클라우드 |
|-----------|------------|------------------|
| 금융 데이터 처리 | 고객 정보, 거래 기록 | 분석용 빅데이터 처리 |
| 제조업 | 공장 자동화 시스템 | IoT 센서 데이터 분석 |
| 교육기관 | 학생 정보 관리 | 온라인 강의 콘텐츠 제공 |



#### ✅ 장점
- **유연성**: 필요에 따라 리소스를 확장하거나 축소 가능
- **보안성**: 민감한 데이터는 내부에서 관리
- **비용 절감**: 퍼블릭 클라우드를 활용해 초기 투자 비용 절감
- **비즈니스 연속성**: 장애 시 클라우드로 자동 전환 가능


하이브리드 클라우드는 단순히 두 환경을 연결하는 것을 넘어서, **통합된 운영과 전략적 자원 배분**이 핵심입니다. 



<br>

## **개발과 운영의 통합(DevOps)**

> 개발 부서와 운영 부서가 협력, 더욱 빠르고 완성도 높은 SW를 만드는 문화.

DevOps에 이목이 쏠리는 배경

1. 개발 방법
2. 클라우드
3. 자동화

개발을 할 대, SV의 전체 그림을 확정한 후 개발하는 게 아닌, 

**SV의 전체 그림 안에서 결정된 부분을 먼저 개발해 배포하는 개발 속도 우선적인 방법**

> app은 테스트 환경을 확장하거나, IT 인프라를 바꿔야 하는 상황이자주 발생하므로,
>
> 개발자 스스로 서버나 스토리지를 생성 + OS 또는 미들웨어의 설정을 바꾸는 경우가 많아짐.

작은 부분으로 쪼개서 구현과 테스트를 반복 -> 전체를 개발하는 방법을 `'agile'`이라고 하는데, 

이런 것들이 가능해진 배경에는 <span style="background-color:#FAFAB4">PaaS SV와 컨테이너 SV가 완비됐기 때문임.</span>

또, 운영진 입장에서도 철저한 자동화로, 운영부서의 리소스 -> 개발부서로 옮길 수 있음.

이를테면 바텀업 형식이네....

<br>

<br>

<br>

<br>

---

## **클라우드 SV를 제공하는 사업자들**

**| Amazon Web SV (AWS)**

- `Amazon EC2(가상 서버)`
- `Amazon S3(클라우드 스토리지)`
- `Amazon RDS(관계형 DB)`
- `Amazon Elastic Container SV(도커 실행/관리)`
- `AWS IoT(IoT와 클라우드 연결)`
- `Amazon Machine Learning(기계학습 기능을 SV)` 등등

기업 사용자의 사내 sys용 VMware의 가상 환경을

AWS에서 구축하고 실행하는 클라우드 SV `VMware Cloud on AWS`도 SV중. ( AWS 물리기반 위에 VMware의 PaaS~ SaaS를 통합한 SV)
